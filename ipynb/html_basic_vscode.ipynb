## 1. HTML이란?

**HTML (HyperText Markup Language)**: 웹페이지의 **구조와 내용**을 정의하는 마크업 언어

| 구분 | 역할 | 비유 |
|------|------|------|
| **HTML** | 웹페이지의 **뼈대**(구조와 내용) | 건물의 골조 |
| **CSS** | 웹페이지의 **디자인**(색상, 크기, 배치) | 건물의 인테리어 |
| **JavaScript** | 웹페이지의 **동작**(클릭, 애니메이션) | 건물의 전기·설비 |

### HTML의 핵심 개념: 태그(Tag)

HTML은 **태그(tag)**로 구성됨. 태그는 꺾쇠괄호(`< >`)로 감싼 키워드

```
<태그이름>내용</태그이름>
 ─────── ──── ────────
 여는 태그  내용  닫는 태그 (슬래시 / 포함)
```

| 용어 | 설명 | 예시 |
|------|------|------|
| **요소(Element)** | 여는 태그 + 내용 + 닫는 태그의 전체 | `<p>안녕하세요</p>` |
| **속성(Attribute)** | 태그에 추가 정보를 부여 | `<a href="url">` 의 `href` |
| **빈 요소(Void Element)** | 닫는 태그가 없는 태그 | `<br>`, `<img>`, `<hr>`, `<input>` |

> **WHATWG HTML Living Standard**: HTML은 더 이상 버전 번호(HTML5 등)가 없고,
> WHATWG가 관리하는 **살아있는 표준(Living Standard)**으로 지속적으로 업데이트됨 (2026년 현재)

````python
# ──────────────────────────────────────────────
# HTML 렌더링 헬퍼 함수 정의
# ──────────────────────────────────────────────
# 이 노트북 전체에서 사용할 함수: HTML 소스코드와 렌더링 결과를 동시에 표시
import html as _html_module
from IPython.display import display, HTML

def show_html(code):
    """HTML 소스코드(위)와 렌더링 결과(아래)를 함께 표시하는 함수"""
    escaped = _html_module.escape(code)   # HTML 태그를 텍스트로 표시하기 위해 이스케이프
    combined = f"""
    <div style="margin-bottom:20px;">
        <div style="background:#1e1e1e; color:#d4d4d4; padding:12px;
                    border-radius:5px 5px 0 0; font-family:'Courier New',monospace;
                    white-space:pre-wrap; font-size:13px; line-height:1.5;">
{escaped}
        </div>
        <div style="padding:12px; border:2px solid #1e1e1e;
                    border-top:none; border-radius:0 0 5px 5px; background:#fff;">
            {code}
        </div>
    </div>
    """
    display(HTML(combined))

# 테스트: 헬퍼 함수가 정상 동작하는지 확인
show_html('<p>HTML 렌더링 테스트 — 소스코드(위)와 결과(아래)가 보이면 성공!</p>')
````

## 2. HTML 문서 기본 구조

모든 HTML 문서는 아래의 **기본 뼈대**를 가짐

```html
<!DOCTYPE html>          ← HTML5 문서임을 선언 (반드시 첫 줄에 위치)
<html lang="ko">         ← 문서 전체를 감싸는 최상위 태그 (lang: 문서 언어)
<head>                    ← 문서의 메타정보 (브라우저에 표시되지 않음)
    <meta charset="UTF-8">           ← 문자 인코딩 (한글 깨짐 방지)
    <meta name="viewport"            ← 모바일 반응형 설정
          content="width=device-width, initial-scale=1.0">
    <title>페이지 제목</title>         ← 브라우저 탭에 표시되는 제목
</head>
<body>                    ← 실제 화면에 표시되는 내용
    <h1>제목</h1>
    <p>본문 내용</p>
</body>
</html>
```

| 태그 | 역할 |
|------|------|
| `<!DOCTYPE html>` | HTML5 문서 선언 (닫는 태그 없음) |
| `<html>` | 문서 전체를 감싸는 루트(root) 요소 |
| `<head>` | 메타정보 영역 (CSS 링크, 문자셋, 제목 등) |
| `<body>` | 화면에 보이는 콘텐츠 영역 |

> **실습 팁**: VSCode에서 `.html` 파일을 만들고 `!` 입력 후 `Tab` 키를 누르면
> 위 기본 구조가 자동 생성됨 (Emmet 기능)

````python
# ──────────────────────────────────────────────
# HTML 문서 기본 구조 시연
# ──────────────────────────────────────────────
# 주의: Jupyter에서는 완전한 HTML 문서를 렌더링하지 않고, body 안의 내용만 렌더링됨
# 아래는 기본 구조를 '소스코드'로 보여주고, body 내용만 실제 렌더링하는 예시

# 1) 전체 구조를 소스코드로 표시 (교육용)
full_html = '''<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>나의 첫 웹페이지</title>
</head>
<body>
    <h1>안녕하세요!</h1>
    <p>나의 첫 번째 HTML 문서입니다.</p>
</body>
</html>'''

print("=== HTML 문서 기본 구조 (소스코드) ===")
print(full_html)

# 2) body 내용만 실제 렌더링
print("\n=== 렌더링 결과 (body 내용) ===")
show_html('''<h1>안녕하세요!</h1>
<p>나의 첫 번째 HTML 문서입니다.</p>''')
````

## 3. 제목 태그 (Heading)

`<h1>` ~ `<h6>`: 문서의 **제목/소제목**을 나타내는 태그
숫자가 작을수록 크고 중요한 제목

| 태그 | 용도 | 크기 |
|------|------|------|
| `<h1>` | 문서의 대제목 (페이지당 1개 권장) | 가장 큼 |
| `<h2>` | 섹션 제목 | ↕ |
| `<h3>` | 하위 섹션 제목 | ↕ |
| `<h4>` ~ `<h6>` | 더 하위 제목 | 가장 작음 |

> **중요**: 제목 태그는 글자 크기를 조절하는 용도가 **아님**
> → 글자 크기는 CSS로 조절하고, 제목 태그는 **문서의 계층 구조**를 나타내는 데 사용
> → h1 → h2 → h3 순서로 사용 (h1 → h3처럼 건너뛰지 않기)

````python
# ──────────────────────────────────────────────
# 제목 태그 h1 ~ h6
# ──────────────────────────────────────────────
show_html('''<h1>h1: 대제목 (페이지당 1개 권장)</h1>
<h2>h2: 섹션 제목</h2>
<h3>h3: 하위 섹션 제목</h3>
<h4>h4: 소제목</h4>
<h5>h5: 더 작은 소제목</h5>
<h6>h6: 가장 작은 제목</h6>''')
````

### 3-1. 단락, 줄바꿈, 수평선

| 태그 | 역할 | 닫는 태그 |
|------|------|----------|
| `<p>` | 단락(paragraph) — 텍스트 블록 | `</p>` 필요 |
| `<br>` | 줄바꿈(line break) — 강제 개행 | 없음 (빈 요소) |
| `<hr>` | 수평선(horizontal rule) — 주제 구분 | 없음 (빈 요소) |

> **주의**: HTML에서는 소스코드의 줄바꿈이나 연속 공백이 **무시됨**
> → 줄바꿈은 `<br>`, 공백은 `&nbsp;`를 사용해야 반영됨

````python
# ──────────────────────────────────────────────
# 단락(p), 줄바꿈(br), 수평선(hr)
# ──────────────────────────────────────────────
show_html('''<p>이것은 첫 번째 단락입니다. p 태그는 텍스트를 하나의 블록으로 묶어줍니다.</p>
<p>이것은 두 번째 단락입니다. 단락 사이에는 자동으로 여백이 생깁니다.</p>

<hr>

<p>위에 수평선(hr)이 보입니다. 주제가 바뀔 때 사용합니다.</p>
<p>줄바꿈 테스트:<br>br 태그로 강제 줄바꿈을 했습니다.<br>한 번 더 줄바꿈.</p>

<p>HTML은 소스코드의       공백을      무시합니다. (연속 공백 → 1칸으로 축소)</p>
<p>공백을&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;넣으려면 &amp;nbsp;를 사용합니다.</p>''')
````

## 4. 텍스트 서식 태그 (Text Formatting)

텍스트에 **의미**와 **시각적 강조**를 부여하는 태그

| 태그 | 의미 | 시각적 효과 |
|------|------|------------|
| `<strong>` | **중요한** 텍스트 (의미적 강조) | 굵게 (bold) |
| `<em>` | **강조**된 텍스트 (의미적 강조) | 기울임 (italic) |
| `<b>` | 시각적으로 굵은 텍스트 (의미 없음) | 굵게 |
| `<i>` | 시각적으로 기울인 텍스트 (의미 없음) | 기울임 |
| `<u>` | 밑줄 텍스트 | 밑줄 |
| `<s>` | 취소선 (더 이상 정확하지 않은 정보) | 취소선 |
| `<del>` | 삭제된 텍스트 (의미적) | 취소선 |
| `<ins>` | 추가된 텍스트 (의미적) | 밑줄 |
| `<mark>` | 하이라이트 (검색 결과 등) | 노란 배경 |
| `<small>` | 부가 정보 (저작권 등) | 작은 글씨 |
| `<sub>` | 아래 첨자 | H₂O의 2 |
| `<sup>` | 위 첨자 | X²의 2 |
| `<code>` | 인라인 코드 | 고정폭 글꼴 |
| `<pre>` | 서식 유지 (공백, 줄바꿈 그대로) | 고정폭 + 서식 보존 |

> **`<strong>` vs `<b>`**: 둘 다 굵게 표시되지만,
> `<strong>`은 "이 텍스트가 중요하다"는 **의미**를 가지고 (스크린 리더가 강조해서 읽음),
> `<b>`는 단순히 시각적 굵기만 적용. **`<strong>` 사용을 권장**

````python
# ──────────────────────────────────────────────
# 텍스트 서식 태그
# ──────────────────────────────────────────────
show_html('''<p><strong>strong: 중요한 텍스트</strong> (의미적 강조, 굵게)</p>
<p><em>em: 강조된 텍스트</em> (의미적 강조, 기울임)</p>
<p><b>b: 굵은 텍스트</b> (시각적, 의미 없음)</p>
<p><i>i: 기울인 텍스트</i> (시각적, 의미 없음)</p>
<p><u>u: 밑줄 텍스트</u></p>
<p><s>s: 취소선 텍스트</s> (더 이상 정확하지 않은 정보)</p>
<p><del>del: 삭제된 텍스트</del> → <ins>ins: 추가된 텍스트</ins></p>
<p><mark>mark: 하이라이트된 텍스트</mark></p>
<p><small>small: 부가 정보 (저작권 표시 등)</small></p>
<p>H<sub>2</sub>O (sub: 아래 첨자) / X<sup>2</sup> (sup: 위 첨자)</p>
<p><code>code: 인라인 코드</code> (예: print("hello"))</p>''')
````

````python
# ──────────────────────────────────────────────
# pre 태그: 공백과 줄바꿈을 그대로 유지
# ──────────────────────────────────────────────
show_html('''<pre>
def hello():
    print("Hello, World!")
    return True
</pre>

<p>pre + code 조합 (코드 블록 표시에 사용):</p>
<pre><code>&lt;h1&gt;제목&lt;/h1&gt;
&lt;p&gt;내용&lt;/p&gt;</code></pre>''')
````

### 4-1. HTML 특수 문자 (Entity)

HTML에서 `<`, `>`, `&` 등은 태그로 인식되므로, **텍스트로 표시하려면 특수 코드** 사용

| 표시 문자 | HTML 코드 | 이름 |
|----------|-----------|------|
| `<` | `&lt;` | less than |
| `>` | `&gt;` | greater than |
| `&` | `&amp;` | ampersand |
| `"` | `&quot;` | quotation mark |
| (공백) | `&nbsp;` | non-breaking space |
| `©` | `&copy;` | copyright |
| `®` | `&reg;` | registered |

````python
# ──────────────────────────────────────────────
# HTML 특수 문자 (Entity)
# ──────────────────────────────────────────────
show_html('''<p>&lt;p&gt; 태그를 텍스트로 표시: &amp;lt; 와 &amp;gt; 사용</p>
<p>저작권: &copy; 2026 My Company</p>
<p>등록상표: HTML&reg;</p>
<p>공백 넣기: 일반공백(1칸) vs &nbsp;&nbsp;&nbsp;nbsp 3개(3칸)</p>''')
````

## 6. 링크 태그 (Anchor)

`<a>` 태그: 다른 페이지, 파일, 위치로 이동하는 **하이퍼링크** 생성

```html
<a href="URL" target="열기방식">링크 텍스트</a>
```

| 속성 | 설명 | 예시 |
|------|------|------|
| `href` | 이동할 URL (필수) | `"https://google.com"` |
| `target` | 링크를 열 위치 | `"_blank"` (새 탭), `"_self"` (현재 탭, 기본값) |
| `title` | 마우스 올렸을 때 표시되는 툴팁 | `"구글 홈페이지"` |
| `download` | 파일 다운로드 링크로 변환 | `download="파일명.pdf"` |

### href 값의 종류

## 10. 시맨틱 태그 (Semantic HTML)

**시맨틱 태그**: 태그 이름만으로 **의미(역할)**를 알 수 있는 태그

`<div>`와 `<span>`은 의미가 없는 범용 컨테이너인 반면,
시맨틱 태그는 **문서의 구조와 의미**를 명확하게 전달

```
┌─────────────────────────────────┐
│           <header>              │  ← 상단 (로고, 제목)
├─────────────────────────────────┤
│            <nav>                │  ← 내비게이션 (메뉴)
├──────────────────┬──────────────┤
│                  │              │
│     <main>       │   <aside>    │  ← 본문 / 사이드바
│  ┌────────────┐  │              │
│  │ <article>  │  │              │
│  └────────────┘  │              │
│  ┌────────────┐  │              │
│  │ <section>  │  │              │
│  └────────────┘  │              │
├──────────────────┴──────────────┤
│           <footer>              │  ← 하단 (저작권, 연락처)
└─────────────────────────────────┘
```

| 태그 | 역할 |
|------|------|
| `<header>` | 문서/섹션의 머리글 (로고, 제목, 내비게이션) |
| `<nav>` | 주요 내비게이션 링크 묶음 |
| `<main>` | 문서의 핵심 콘텐츠 (**페이지당 1개**) |
| `<article>` | 독립적으로 배포 가능한 콘텐츠 (뉴스 기사, 블로그 글) |
| `<section>` | 주제별로 묶은 콘텐츠 그룹 (제목 포함 권장) |
| `<aside>` | 본문과 간접적으로 관련된 부가 콘텐츠 (사이드바) |
| `<footer>` | 문서/섹션의 바닥글 (저작권, 연락처) |
| `<search>` | 검색/필터링 기능을 감싸는 영역 (2025+ 최신) |
| `<figure>` | 독립적 콘텐츠 (이미지, 다이어그램 등) + `<figcaption>` |
| `<time>` | 날짜/시간 (기계 판독 가능) |
| `<address>` | 연락처 정보 |
| `<mark>` | 강조/하이라이트 텍스트 |

### 왜 시맨틱 태그를 사용해야 하는가?

| 이유 | 설명 |
|------|------|
| **접근성** | 스크린 리더가 페이지 구조를 이해하고 올바르게 읽어줌 |
| **SEO** | 검색엔진이 콘텐츠의 중요도와 구조를 파악 |
| **유지보수** | 코드만 봐도 각 영역의 역할을 즉시 파악 가능 |
| **표준 준수** | WCAG 2.2+ 웹 접근성 지침 충족 |

> **`<div>` vs 시맨틱 태그**:
> `<div>`는 의미 없는 범용 컨테이너 → CSS 레이아웃용으로만 사용
> 의미를 나타낼 수 있는 경우 항상 시맨틱 태그를 우선 사용


| 속성 | 위치 | 설명 |

| `popovertarget` | 버튼 | 연결할 팝오버의 id |
| `popovertargetaction` | 버튼 | `"toggle"` (기본), `"show"`, `"hide"` |

> Popover는 **비모달**: 배경이 비활성화되지 않으며, 바깥 클릭/Esc로 닫힘
> 모든 주요 브라우저에서 지원 (2025 Baseline)

````python
# ──────────────────────────────────────────────
# Popover API 소스코드 (실제 .html 파일에서 사용)
# ──────────────────────────────────────────────
# Popover는 Jupyter에서 동작하지 않을 수 있음 — 소스코드로 안내

popover_example = '''<!-- Popover 기본 사용법 -->
<button popovertarget="my-popover">정보 보기</button>
<div id="my-popover" popover>
    <p>이것은 팝오버 내용입니다!</p>
    <p>바깥을 클릭하거나 Esc를 누르면 닫힙니다.</p>
</div>

<!-- Popover 수동 모드 (manual) -->
<button popovertarget="manual-pop" popovertargetaction="show">열기</button>
<button popovertarget="manual-pop" popovertargetaction="hide">닫기</button>
<div id="manual-pop" popover="manual">
    <p>수동 모드: 닫기 버튼을 눌러야 닫힙니다.</p>
</div>

<!-- Invoker Commands (2026 신기능) -->
<button commandfor="my-dialog" command="show-modal">모달 열기</button>
<dialog id="my-dialog">
    <p>commandfor/command로 JavaScript 없이 모달 제어!</p>
    <button commandfor="my-dialog" command="close">닫기</button>
</dialog>'''

print("=== Popover API / Invoker Commands 소스코드 ===")
print(popover_example)
````

## 13. `<head>` 영역의 메타 태그

`<head>` 안에 들어가는 태그들: 브라우저와 검색엔진에 **문서 정보를 제공**

| 태그 | 용도 |
|------|------|
| `<meta charset="UTF-8">` | 문자 인코딩 (한글 깨짐 방지, **필수**) |
| `<meta name="viewport" ...>` | 모바일 반응형 설정 (**필수**) |
| `<meta name="description" ...>` | 검색 결과에 표시되는 페이지 설명 (SEO) |
| `<meta name="keywords" ...>` | 검색 키워드 (현재는 SEO 효과 미미) |
| `<meta name="author" ...>` | 작성자 |
| `<title>` | 브라우저 탭 제목 + 검색 결과 제목 (**필수**) |
| `<link rel="stylesheet" ...>` | 외부 CSS 파일 연결 |
| `<link rel="icon" ...>` | 파비콘 (탭 아이콘) |
| `<script src="...">` | JavaScript 파일 연결 |
| `<style>` | 내부 CSS 작성 |

### Open Graph 메타 태그 (SNS 공유용)

```html
<meta property="og:title" content="페이지 제목">

<meta property="og:url" content="https://example.com/page">
```

## 14. 블록(Block) 요소 vs 인라인(Inline) 요소

HTML 요소는 **블록**과 **인라인** 두 가지 표시 유형으로 나뉨

| 구분 | 블록(Block) 요소 | 인라인(Inline) 요소 |
|------|----------------|-------------------|
| **너비** | 부모 요소의 **전체 너비** 차지 | 콘텐츠만큼만 차지 |
| **배치** | 항상 **새 줄**에서 시작 | 같은 줄에 **나란히** 배치 |
| **예시** | `<div>`, `<p>`, `<h1>`~`<h6>`, `<ul>`, `<table>`, `<form>`, `<header>`, `<section>` | `<span>`, `<a>`, `<strong>`, `<em>`, `<img>`, `<input>`, `<code>`, `<br>` |

```
블록 요소:                    인라인 요소:
┌──────────────────────┐      텍스트 [a링크] [strong] 텍스트
│     <div> 블록 1      │      → 한 줄 안에 나란히 배치
├──────────────────────┤
│     <p> 블록 2        │
└──────────────────────┘
→ 각각 한 줄 전체 차지
```

````python
# ──────────────────────────────────────────────
# 블록 vs 인라인 요소 비교
# ──────────────────────────────────────────────
show_html('''<h3>블록(Block) 요소 — 한 줄 전체 차지</h3>
<div style="background:#AED6F1; padding:8px; margin:4px 0;">div 블록 1</div>
<p style="background:#A9DFBF; padding:8px; margin:4px 0;">p 블록 2</p>
<div style="background:#F9E79F; padding:8px; margin:4px 0;">div 블록 3</div>

<h3>인라인(Inline) 요소 — 같은 줄에 나란히</h3>
<p>
    일반 텍스트
    <a href="#" style="background:#AED6F1; padding:2px 6px;">a 링크</a>
    <strong style="background:#A9DFBF; padding:2px 6px;">strong 강조</strong>
    <span style="background:#F9E79F; padding:2px 6px;">span 텍스트</span>
    <code style="background:#D7BDE2; padding:2px 6px;">code 코드</code>
    일반 텍스트 계속
</p>''')
````

## 15. 전역 속성 (Global Attributes)

**모든 HTML 태그에 사용 가능**한 공통 속성

| 속성 | 설명 | 예시 |
|------|------|------|
| `id` | 요소의 고유 식별자 (페이지 내 유일) | `id="header"` |
| `class` | 요소의 분류 (CSS/JS에서 그룹 선택, 여러 개 가능) | `class="box red"` |
| `style` | 인라인 CSS 스타일 | `style="color:red;"` |
| `title` | 마우스 올렸을 때 툴팁 텍스트 | `title="설명"` |
| `hidden` | 요소를 숨김 | `<p hidden>안 보임</p>` |
| `lang` | 요소의 언어 지정 | `lang="en"` |
| `data-*` | 사용자 정의 데이터 저장 | `data-user-id="123"` |
| `tabindex` | Tab 키 이동 순서 | `tabindex="1"` |
| `contenteditable` | 내용 편집 가능 | `contenteditable="true"` |
| `draggable` | 드래그 가능 여부 | `draggable="true"` |
| `inert` | 비활성화 (2025+ 신기능) | `<div inert>...</div>` |

### id vs class

| 구분 | `id` | `class` |
|------|------|---------|
| 고유성 | 페이지 내 **유일** | **중복 가능** |
| 개수 | 요소당 **1개** | 요소당 **여러 개** 가능 |
| CSS 선택 | `#id이름` | `.class이름` |
| 용도 | 특정 요소 지정 (앵커, JS) | 스타일 그룹핑 |

````python
# ──────────────────────────────────────────────
# 전역 속성 활용 예시
# ──────────────────────────────────────────────
show_html('''<p id="intro" class="highlight" title="마우스를 올려보세요!">
    id="intro", class="highlight", title 속성이 있는 단락 (마우스 올려보세요)
</p>

<p style="color:blue; font-size:18px; background:#EBF5FB; padding:8px;">
    style 속성으로 인라인 CSS 적용 (파란색, 18px, 배경색)
</p>

<p hidden>이 텍스트는 hidden 속성으로 숨겨져 보이지 않습니다.</p>

<p contenteditable="true"
   style="border:1px dashed #999; padding:8px; background:#FFF9C4;">
    contenteditable="true" — 이 텍스트를 직접 클릭하여 편집할 수 있습니다! (수정해 보세요)
</p>

<p data-category="html" data-level="beginner">
    data-* 속성: data-category="html", data-level="beginner"<br>
    <small>(화면에는 표시되지 않지만 JavaScript에서 접근 가능)</small>
</p>''')
````

## 16. HTML 주석 (Comment)


- 코드 설명, 임시 비활성화, 개발 메모 등에 사용
- 주석은 브라우저 **개발자 도구(F12)**에서 확인 가능하므로 민감한 정보 금지

````python
# ──────────────────────────────────────────────
# HTML 주석 (화면에 표시되지 않음)
# ──────────────────────────────────────────────
comment_html = '''<!-- 이 주석은 화면에 보이지 않습니다 -->
<p>이 텍스트만 화면에 표시됩니다.</p>

<!-- 아래 코드는 임시로 비활성화 -->
<!-- <p style="color:red;">이 빨간 텍스트는 주석 처리되어 보이지 않음</p> -->

<p>주석은 소스코드에만 존재합니다.</p>'''

print("=== HTML 소스코드 ===")
print(comment_html)
print()
show_html(comment_html)
````

## 📋 HTML 기초 태그 요약 (Quick Reference)

| 카테고리 | 태그 | 설명 |
|---------|------|------|
| **문서 구조** | `<!DOCTYPE html>`, `<html>`, `<head>`, `<body>` | HTML 문서의 기본 뼈대 |
| **메타 정보** | `<meta>`, `<title>`, `<link>`, `<style>`, `<script>` | head 안의 문서 정보 |
| **제목** | `<h1>` ~ `<h6>` | 문서의 계층적 제목 |
| **텍스트** | `<p>`, `<br>`, `<hr>` | 단락, 줄바꿈, 수평선 |
| **서식** | `<strong>`, `<em>`, `<mark>`, `<code>`, `<pre>` | 텍스트 의미·서식 |
| **목록** | `<ul>`, `<ol>`, `<li>`, `<dl>`, `<dt>`, `<dd>` | 순서 있는/없는/정의 목록 |
| **링크** | `<a href="URL">` | 하이퍼링크 |
| **이미지** | `<img src="" alt="">` | 이미지 삽입 |
| **테이블** | `<table>`, `<thead>`, `<tbody>`, `<tr>`, `<th>`, `<td>` | 표 |
| **폼** | `<form>`, `<input>`, `<select>`, `<textarea>`, `<button>` | 사용자 입력 |
| **시맨틱** | `<header>`, `<nav>`, `<main>`, `<article>`, `<section>`, `<aside>`, `<footer>` | 의미 있는 구조 태그 |
| **미디어** | `<audio>`, `<video>`, `<picture>`, `<iframe>` | 멀티미디어 삽입 |
| **최신 기능** | `<details>`, `<dialog>`, `popover`, `inert` | JS 없이 인터랙티브 UI |
| **범용** | `<div>` (블록), `<span>` (인라인) | 의미 없는 컨테이너 |

### 핵심 원칙 요약

| 원칙 | 설명 |
|------|------|
| **구조와 의미** | HTML은 구조를, CSS는 디자인을, JS는 동작을 담당 |
| **시맨틱 우선** | `<div>` 대신 의미 있는 시맨틱 태그를 우선 사용 |
| **접근성** | `alt`, `<label>`, 시맨틱 태그로 모든 사용자가 이용 가능하도록 |
| **제목 계층** | `<h1>` → `<h2>` → `<h3>` 순서를 지키고, 건너뛰지 않기 |
| **Living Standard** | HTML은 버전이 아닌 지속적으로 업데이트되는 살아있는 표준 |
