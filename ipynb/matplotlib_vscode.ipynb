## 0. matplotlib이란?

파이썬에서 **그래프와 차트를 그리기 위한 가장 기본적인 시각화 라이브러리**  
거의 모든 파이썬 시각화 도구(seaborn, pandas plot 등)가 matplotlib 위에 구축됨

### 핵심 구조

```
Figure (전체 그림 = 캔버스)
├── Axes (그래프 영역 = 좌표축)
│   ├── Title (제목)
│   ├── X-axis (x축) ─ xlabel, xticks
│   ├── Y-axis (y축) ─ ylabel, yticks
│   ├── Legend (범례)
│   ├── Grid (격자)
│   └── Plot 데이터 (선, 점, 막대 등)
├── Axes (두 번째 그래프)
└── ...
```

### 주요 그래프 유형

| 그래프 | 함수 | 용도 |
|--------|------|------|
| 산점도 | `plt.scatter()` | 두 변수의 관계 (상관관계) |
| 라인 플롯 | `plt.plot()` | 시간에 따른 변화 (추세) |
| 막대 그래프 | `plt.bar()` | 범주별 비교 |
| 파이 차트 | `plt.pie()` | 비율/구성 |
| 히스토그램 | `plt.hist()` | 데이터 분포 |
| 히트맵 | `plt.pcolormesh()` | 2차원 데이터 강도 |

## 2. 라이브러리 불러오기 & 한글 폰트 설정

matplotlib에서 한글을 사용하려면 **폰트 설정이 필수**  
VSCode 로컬 환경에서는 OS에 설치된 폰트를 직접 지정

| OS | 한글 폰트 | 설정값 |
|----|----------|--------|
| Windows | 맑은 고딕 | `'Malgun Gothic'` |
| macOS | Apple Gothic | `'AppleGothic'` |
| Linux | 나눔고딕 | `'NanumGothic'` (별도 설치 필요) |

````python
# ──────────────────────────────────────────────
# 필수 라이브러리 불러오기 & 한글 폰트 설정
# ──────────────────────────────────────────────
import matplotlib.pyplot as plt   # matplotlib: 그래프 시각화
import numpy as np                # numpy: 수치 연산
import pandas as pd               # pandas: 데이터 처리
import platform                   # OS 확인용

# --- 한글 폰트 설정 (OS 자동 감지) ---
current_os = platform.system()
if current_os == 'Windows':
    plt.rc('font', family='Malgun Gothic')   # Windows: 맑은 고딕
elif current_os == 'Darwin':                  # macOS
    plt.rc('font', family='AppleGothic')
else:                                         # Linux
    plt.rc('font', family='NanumGothic')

# 마이너스 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 버전 확인
import matplotlib
print(f"matplotlib 버전: {matplotlib.__version__}")
print(f"OS: {current_os}")
````

## 3. 그래프의 두 가지 방식

matplotlib에는 그래프를 그리는 **두 가지 스타일**이 있음  
상황에 따라 적절한 방식을 선택

| 비교 | 상태기반 방식 (plt) | 객체지향 방식 (fig, ax) |
|------|-------------------|----------------------|
| **코드** | `plt.plot()`, `plt.title()` | `ax.plot()`, `ax.set_title()` |
| **장점** | 간단, 빠른 시각화 | 세밀한 제어, 서브플롯 |
| **적합** | 단일 그래프 | 여러 그래프, 복잡한 레이아웃 |
| **비유** | 자동 모드 | 수동 모드 |

````python
# ──────────────────────────────────────────────
# 방법 1: 상태기반 방식 (plt 직접 사용)
#   간단한 그래프에 적합. plt에 명령을 순서대로 쌓아감
# ──────────────────────────────────────────────
x = [50, 60, 70, 80, 90]
y = [20, 50, 70, 90, 100]

# 1) 그래프 크기 설정
plt.figure(figsize=(5, 4))
# 2) 데이터 그리기
plt.scatter(x, y, marker='o', color='blue', s=100, alpha=0.7)
# 3) 축 이름과 제목
plt.xlabel('영어 점수', fontsize=11)
plt.ylabel('수학 점수', fontsize=11)
plt.title('영어와 수학 점수 관계 (상태기반)', fontsize=14)
# 4) 격자와 범례
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(labels=['영어-수학'], loc='upper left')
# 5) 여백 자동 조정 후 표시
plt.tight_layout()
plt.show()
````

````python
# ──────────────────────────────────────────────
# 방법 2: 객체지향 방식 (fig, ax 사용)
#   복잡한 그래프에 적합. fig=캔버스, ax=그래프 영역 객체
# ──────────────────────────────────────────────
x = [50, 60, 70, 80, 90]
y = [20, 50, 70, 90, 100]

# 1) Figure(캔버스)와 Axes(그래프 영역) 생성
fig, ax = plt.subplots(1, 1, figsize=(5, 4))
#   subplots(행, 열) → 그래프 개수 지정 가능

# 2) ax 객체를 통해 데이터 그리기
ax.scatter(x, y, marker='o', color='red', s=100, alpha=0.7)
# 3) ax 객체를 통해 꾸미기 (set_ 접두사 사용)
ax.set_xlabel('영어 점수', fontsize=11)
ax.set_ylabel('수학 점수', fontsize=11)
ax.set_title('영어와 수학 점수 관계 (객체지향)', fontsize=14)
ax.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
````

## 5. 산점도 (Scatter Plot)

**두 변수 간의 관계(상관관계)**를 시각적으로 확인할 때 사용  
점의 위치, 색상, 크기로 데이터 특성을 표현

### `plt.scatter()` 주요 매개변수

| 매개변수 | 설명 | 예시 |
|---------|------|------|
| `x, y` | x축, y축 데이터 | 리스트 또는 배열 |
| `marker` | 점 모양 | `'o'`, `'^'`, `'s'` |
| `s` | 점 크기 | `100` |
| `c` / `color` | 점 색상 | `'red'`, `'#FF0000'` |
| `alpha` | 투명도 (0~1) | `0.7` |
| `edgecolors` | 테두리 색 | `'black'` |
| `label` | 범례 텍스트 | `'A반'` |

````python
# ──────────────────────────────────────────────
# 산점도: 1개 집단
# ──────────────────────────────────────────────
영어 = [51, 62, 43, 74, 85, 56, 67, 98, 49, 68]
수학 = [53, 65, 42, 66, 78, 76, 74, 98, 88, 76]

plt.figure(figsize=(8, 5))
plt.scatter(영어, 수학,
            marker='o',              # 마커 모양: 원
            s=100,                   # 마커 크기
            c='red',                 # 마커 색상
            alpha=0.7,               # 투명도
            edgecolors='black',      # 테두리 색
            linewidths=1,            # 테두리 굵기
            label='학생별 영어-수학 점수')

plt.xlabel('영어 점수', fontsize=12)
plt.ylabel('수학 점수', fontsize=12)
plt.title('영어와 수학 점수의 상관관계', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# ──────────────────────────────────────────────
# 산점도: 복수 집단 비교
# ──────────────────────────────────────────────
영어_A반 = [51, 62, 43, 74, 85, 56, 67, 98, 49, 68]
수학_A반 = [53, 65, 42, 66, 78, 76, 74, 98, 88, 76]
영어_B반 = [71, 77, 87, 74, 88, 86, 87, 88, 89, 78]
수학_B반 = [73, 65, 72, 86, 78, 76, 74, 88, 78, 86]

plt.figure(figsize=(8, 5))

# 각 집단을 서로 다른 색상/마커로 구분
plt.scatter(영어_A반, 수학_A반,
            marker='o', s=120, c='red', alpha=0.7,
            label='A반', edgecolors='black')

plt.scatter(영어_B반, 수학_B반,
            marker='^', s=120, c='blue', alpha=0.7,
            label='B반', edgecolors='black')

plt.legend(fontsize=11, loc='upper left')
plt.xlabel('영어 점수', fontsize=12)
plt.ylabel('수학 점수', fontsize=12)
plt.title('A반과 B반의 영어-수학 점수 비교', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

## 6. 라인 플롯 (Line Plot)

**시간에 따른 변화(추세)**를 표현할 때 사용  
연도별, 월별, 일별 데이터 등 **시계열 시각화**에 핵심

### `plt.plot()` 주요 매개변수

| 매개변수 | 설명 | 예시 |
|---------|------|------|
| `marker` | 데이터 포인트 마커 | `'o'`, `'s'`, `'^'` |
| `linestyle` | 선 스타일 | `'-'`, `'--'`, `'-.'`, `':'` |
| `linewidth` | 선 굵기 | `2`, `3` |
| `color` | 선 색상 | `'blue'` |
| `markerfacecolor` | 마커 내부 색 | `'white'` |
| `markersize` | 마커 크기 | `10` |
| `label` | 범례 텍스트 | `'수학 평균'` |

````python
# ──────────────────────────────────────────────
# 라인 플롯: 1개 변수
# ──────────────────────────────────────────────
연도 = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
수학_평균 = [30, 35, 44, 50, 51, 67, 70, 73]

plt.figure(figsize=(9, 5))

plt.plot(연도, 수학_평균,
         marker='o',                 # 데이터 포인트 표시
         linestyle='-',              # 실선
         linewidth=3,                # 선 굵기
         color='blue',               # 선 색상
         alpha=0.8,                  # 투명도
         markerfacecolor='white',    # 마커 내부 흰색
         markeredgecolor='blue',     # 마커 테두리 색
         markeredgewidth=2,          # 마커 테두리 굵기
         markersize=10,              # 마커 크기
         label='수학 평균')

plt.xlabel('연도', fontsize=12)
plt.ylabel('평균 점수', fontsize=12)
plt.title('연도별 수학 평균 점수 변화 추이', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
````

````python
# ──────────────────────────────────────────────
# 라인 플롯: 복수 변수 비교
# ──────────────────────────────────────────────
연도 = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
영어_평균 = [50, 60, 40, 50, 60, 40, 50, 60]
수학_평균 = [30, 35, 44, 50, 51, 67, 70, 73]

plt.figure(figsize=(9, 5))

# 선 스타일을 달리하여 구분
plt.plot(연도, 영어_평균,
         marker='o', linestyle='-', linewidth=3,
         color='red', alpha=0.8, label='영어 평균')

plt.plot(연도, 수학_평균,
         marker='s', linestyle='--', linewidth=3,
         color='blue', alpha=0.8, label='수학 평균')

plt.xlabel('연도', fontsize=12)
plt.ylabel('평균 점수', fontsize=12)
plt.title('연도별 교과목 평균 점수 추이 비교', fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11, loc='lower right')
plt.tight_layout()
plt.show()
````

## 7. 막대 그래프 (Bar Plot)

**범주(카테고리)별 값을 비교**할 때 사용  
막대 위에 값을 표시하면 정확한 수치를 함께 전달 가능

### `plt.bar()` 주요 매개변수

| 매개변수 | 설명 | 예시 |
|---------|------|------|
| `x` | 범주(x축 위치) | `['A', 'B', 'C']` |
| `height` | 막대 높이 | `[70, 60, 75]` |
| `width` | 막대 너비 | `0.6` |
| `color` | 막대 색상 | `'skyblue'` |
| `edgecolor` | 테두리 색상 | `'black'` |

> 가로 막대: `plt.barh()` 사용

````python
# ──────────────────────────────────────────────
# 막대 그래프: 1개 집단 + 값 표시
# ──────────────────────────────────────────────
반 = ['A반', 'B반', 'C반']
영어평균 = [70, 60, 75]

plt.figure(figsize=(8, 5))

# 막대 생성
bars = plt.bar(반, 영어평균,
               width=0.6,               # 막대 너비
               color='skyblue',          # 막대 색상
               edgecolor='black',        # 테두리 색상
               linewidth=1.5,            # 테두리 굵기
               alpha=0.8,                # 투명도
               label='영어 평균')

# 막대 위에 값 표시
for bar, value in zip(bars, 영어평균):
    plt.text(
        bar.get_x() + bar.get_width() / 2,   # x위치: 막대 중앙
        value + 1,                             # y위치: 막대 위 약간 위
        str(value),                            # 표시할 텍스트
        ha='center', va='bottom',              # 가운데/아래 정렬
        fontweight='bold', fontsize=11)

plt.xlabel('반', fontsize=12)
plt.ylabel('평균 점수', fontsize=12)
plt.title('반별 영어 평균 점수 비교', fontsize=15)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.ylim(0, 100)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# ──────────────────────────────────────────────
# 막대 그래프: 복수 집단 (그룹화된 막대)
# ──────────────────────────────────────────────
반 = ['A반', 'B반', 'C반']
영어평균 = [50, 60, 40]
수학평균 = [40, 60, 70]

# x축 위치 계산: 막대를 나란히 놓기 위해
x_locs = np.arange(len(반))     # [0, 1, 2]
bar_width = 0.3                   # 각 막대의 너비

plt.figure(figsize=(8, 5))

# 영어 막대 (왼쪽 배치)
bars1 = plt.bar(x_locs - bar_width/2, 영어평균, bar_width,
                label='영어', color='red', edgecolor='black')
# 수학 막대 (오른쪽 배치)
bars2 = plt.bar(x_locs + bar_width/2, 수학평균, bar_width,
                label='수학', color='blue', edgecolor='black')

# 막대 위에 값 표시 (함수 정의하여 재사용)
def add_labels(bars, values):
    """막대 위에 값을 표시하는 헬퍼 함수"""
    for bar, value in zip(bars, values):
        plt.text(bar.get_x() + bar.get_width()/2,
                 value + 1, str(value),
                 ha='center', va='bottom',
                 fontsize=10, fontweight='bold')

add_labels(bars1, 영어평균)
add_labels(bars2, 수학평균)

plt.xlabel('반', fontsize=12)
plt.ylabel('평균 점수', fontsize=12)
plt.title('반별 교과목 평균 점수 비교', fontsize=15)
plt.xticks(x_locs, 반)           # x축 눈금을 반 이름으로
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.ylim(0, 100)
plt.legend(title='교과목')
plt.tight_layout()
plt.show()
````

## 8. 파이 차트 (Pie Chart)

**전체 대비 각 항목의 비율(구성)**을 표현할 때 사용  
`wedgeprops`의 `width`를 조절하면 **도넛 차트**도 가능

### `plt.pie()` 주요 매개변수

| 매개변수 | 설명 | 예시 |
|---------|------|------|
| `x` | 각 조각의 크기 | `[4, 6, 3]` |
| `labels` | 각 조각의 이름 | `['국어', '수학', '영어']` |
| `autopct` | 비율 표시 형식 | `'%.1f%%'` → `25.0%` |
| `startangle` | 시작 각도 | `90` (12시 방향) |
| `counterclock` | 반시계 방향 | `False` → 시계 방향 |
| `explode` | 조각 분리 거리 | `(0.1, 0, 0)` |
| `wedgeprops` | 조각 속성 | `{'width': 0.7}` → 도넛 |

````python
# ──────────────────────────────────────────────
# 파이 차트 (도넛 스타일)
# ──────────────────────────────────────────────
과목 = ['국어', '수학', '영어', '과학', '사회']
시간 = [4, 6, 3, 2, 1]   # 주당 공부 시간

plt.figure(figsize=(8, 6))

# 컬러맵에서 색상 자동 추출
cmap = plt.get_cmap('Set3')   # 또는 'tab10', 'Pastel1' 등
colors = [cmap(i) for i in range(len(과목))]

plt.pie(시간,
        labels=과목,                         # 각 조각 이름
        autopct='%.1f%%',                    # 비율 소수점 1자리
        startangle=90,                       # 12시 방향 시작
        counterclock=False,                  # 시계 방향
        colors=colors,                       # 색상
        wedgeprops={'width': 0.7,            # 도넛 두께 (1.0이면 일반 파이)
                    'edgecolor': 'white',    # 조각 경계 색
                    'linewidth': 2},         # 경계 두께
        textprops={'fontsize': 12},          # 텍스트 크기
        shadow=True,                         # 그림자 효과
        explode=(0.1, 0, 0, 0, 0))           # 국어 조각 분리

plt.title('과목별 주당 학습 시간 비율', fontsize=15)

# 범례 (오른쪽에 배치)
legend_labels = [f'{과} - {시}시간' for 과, 시 in zip(과목, 시간)]
plt.legend(legend_labels, title='과목별 시간',
           loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
````

## 9. 히스토그램 (Histogram)

**데이터의 분포(빈도)**를 시각화할 때 사용  
막대 그래프와 비슷하지만, **연속 데이터의 구간별 빈도**를 나타냄

| 매개변수 | 설명 | 예시 |
|---------|------|------|
| `bins` | 구간 수 | `10`, `20` |
| `edgecolor` | 막대 경계 색 | `'black'` |
| `density` | 확률밀도로 표시 | `True` |

````python
# ──────────────────────────────────────────────
# 히스토그램: 데이터 분포 확인
# ──────────────────────────────────────────────
rng = np.random.default_rng(42)
scores = rng.normal(loc=70, scale=10, size=200)   # 평균 70, 표준편차 10, 200명

plt.figure(figsize=(8, 5))

plt.hist(scores,
         bins=15,                     # 구간 수
         color='skyblue',             # 막대 색상
         edgecolor='black',           # 경계 색
         alpha=0.7,                   # 투명도
         label='점수 분포')

# 평균선 표시
plt.axvline(x=np.mean(scores), color='red', linestyle='--',
            linewidth=2, label=f'평균: {np.mean(scores):.1f}')

plt.xlabel('점수', fontsize=12)
plt.ylabel('빈도 (학생 수)', fontsize=12)
plt.title('학생 점수 분포 (200명)', fontsize=15)
plt.legend(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

## 10. 히트맵 (Heatmap)

**2차원 데이터의 크기를 색상 강도로 표현**하는 그래프  
상관관계 행렬, 교차표, 시간-카테고리별 통계 등에 활용

| 함수 | 설명 |
|------|------|
| `plt.pcolormesh()` | 2차원 배열을 색상 격자로 표시 |
| `plt.colorbar()` | 색상 막대(범례) 추가 |
| `cmap` | 색상 맵 (`'Oranges'`, `'Blues'`, `'YlOrRd'` 등) |

````python
# ──────────────────────────────────────────────
# 히트맵: 샘플 데이터로 구현
# ──────────────────────────────────────────────

# 샘플 데이터: 과목 × 분기별 평균 점수
과목명 = ['국어', '수학', '영어', '과학', '사회']
분기명 = ['1분기', '2분기', '3분기', '4분기']

rng = np.random.default_rng(7)
data = rng.integers(50, 100, size=(len(과목명), len(분기명)))

# 히트맵 생성
plt.figure(figsize=(8, 5))
plt.pcolormesh(data, cmap='YlOrRd', edgecolors='white', linewidth=1)

# 축 라벨 위치 조정 (셀 중앙에 배치)
col_pos = np.arange(0.5, len(분기명), 1)
row_pos = np.arange(0.5, len(과목명), 1)
plt.xticks(col_pos, 분기명, fontsize=11)
plt.yticks(row_pos, 과목명, fontsize=11)

# 셀 안에 값 표시
for i, y_pos in enumerate(row_pos):
    for j, x_pos in enumerate(col_pos):
        value = data[i, j]
        plt.text(x_pos, y_pos, str(value),
                 ha='center', va='center',
                 fontsize=12, fontweight='bold', color='black')

plt.colorbar(label='점수')   # 색상 막대
plt.title('과목 × 분기별 평균 점수', fontsize=15)
plt.tight_layout()
plt.show()
````

````python
# ──────────────────────────────────────────────
# 사용 가능한 컬러맵(cmap) 목록 확인
# ──────────────────────────────────────────────
# 순차형: 'Blues', 'Greens', 'Oranges', 'Reds', 'YlOrRd'
# 발산형: 'RdBu', 'coolwarm', 'seismic'
# 범주형: 'Set1', 'Set2', 'Set3', 'tab10', 'Pastel1'
print("사용 가능한 컬러맵 수:", len(plt.colormaps()))
print("주요 컬러맵:", ['Blues', 'Oranges', 'YlOrRd', 'coolwarm', 'viridis', 'Set3'])
````

## 11. 서브플롯 (Subplots) — 여러 그래프 한 번에

하나의 Figure(캔버스) 안에 **여러 개의 그래프를 배치**  
데이터를 다각도로 비교할 때 유용

```
fig, axes = plt.subplots(행, 열)

subplots(1, 2):              subplots(2, 2):
┌────────┬────────┐          ┌────────┬────────┐
│ ax[0]  │ ax[1]  │          │ax[0,0] │ax[0,1] │
└────────┴────────┘          ├────────┼────────┤
  → 1행 2열                  │ax[1,0] │ax[1,1] │
                             └────────┴────────┘
                               → 2행 2열
```

````python
# ──────────────────────────────────────────────
# 서브플롯: 2x2 = 4개 그래프를 한 화면에
# ──────────────────────────────────────────────
rng = np.random.default_rng(42)

# Figure와 4개의 Axes 생성
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# --- 좌상: 산점도 ---
x = rng.normal(50, 10, 50)
y = x + rng.normal(0, 5, 50)
axes[0, 0].scatter(x, y, c='red', alpha=0.6, edgecolors='black')
axes[0, 0].set_title('산점도')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')

# --- 우상: 라인 플롯 ---
years = range(2015, 2025)
values = np.cumsum(rng.integers(-5, 10, len(years)))
axes[0, 1].plot(years, values, marker='o', color='blue', linewidth=2)
axes[0, 1].set_title('라인 플롯')
axes[0, 1].set_xlabel('연도')

# --- 좌하: 막대 그래프 ---
categories = ['A', 'B', 'C', 'D']
vals = [45, 78, 56, 90]
axes[1, 0].bar(categories, vals, color=['coral', 'skyblue', 'lightgreen', 'gold'],
               edgecolor='black')
axes[1, 0].set_title('막대 그래프')
axes[1, 0].set_ylabel('값')

# --- 우하: 히스토그램 ---
data = rng.normal(100, 15, 300)
axes[1, 1].hist(data, bins=20, color='plum', edgecolor='black', alpha=0.7)
axes[1, 1].set_title('히스토그램')
axes[1, 1].set_xlabel('값')
axes[1, 1].set_ylabel('빈도')

# 전체 제목과 여백
fig.suptitle('주요 그래프 유형 한눈에 보기', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
````

## 12. pandas DataFrame과 연동

pandas DataFrame은 `.plot()` 메서드를 내장하고 있어  
**별도의 plt 호출 없이도 간편하게 시각화** 가능

| DataFrame 메서드 | 그래프 |
|-----------------|--------|
| `df.plot()` | 라인 플롯 (기본) |
| `df.plot.bar()` | 막대 그래프 |
| `df.plot.barh()` | 수평 막대 |
| `df.plot.scatter()` | 산점도 |
| `df.plot.pie()` | 파이 차트 |
| `df.plot.hist()` | 히스토그램 |

````python
# ──────────────────────────────────────────────
# pandas DataFrame → 라인 플롯
# ──────────────────────────────────────────────

# 샘플 DataFrame 생성
df = pd.DataFrame({
    '사고건수':  [320, 310, 305, 290, 285, 270, 260, 250, 240, 235],
    '중상자수': [150, 145, 140, 135, 130, 125, 120, 115, 110, 105]
}, index=range(2015, 2025))
df.index.name = '연도'

print("=== 데이터 확인 ===")
display(df)

# pandas의 .plot() 메서드로 바로 시각화
df.plot(marker='o', linewidth=2, alpha=0.8,
        markersize=8, figsize=(9, 5))

plt.title('연도별 교통사고 통계 추이', fontsize=15)
plt.xlabel('연도', fontsize=12)
plt.ylabel('건수', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='항목', fontsize=11)
plt.tight_layout()
plt.show()
````

````python
# ──────────────────────────────────────────────
# pandas DataFrame → 막대 그래프
# ──────────────────────────────────────────────

# 샘플: 사고유형별 통계
df_type = pd.DataFrame({
    '사고건수': [3.2, 4.1, 2.8, 5.0, 3.5],
    '중상자수': [1.5, 2.3, 1.2, 2.8, 1.7]
}, index=['보행노인', '보행어린이', '자전거', '스쿨존', '기타'])

# DataFrame의 .plot.bar() 사용
ax = df_type.plot.bar(
    color=['tomato', 'steelblue'],
    alpha=0.7,
    edgecolor='black',
    figsize=(9, 5)
)

plt.title('사고 유형별 평균 통계 비교', fontsize=15)
plt.xlabel('사고 유형', fontsize=12)
plt.ylabel('평균 수', fontsize=12)
plt.legend(title='항목', fontsize=11)
plt.xticks(rotation=0)            # x축 레이블 가로 표시
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

````python
# ──────────────────────────────────────────────
# pandas DataFrame → 파이 차트
# ──────────────────────────────────────────────

# 사고유형별 사고건수 합계 (샘플)
df_pie = pd.DataFrame({
    '사고건수': [1200, 1500, 800, 1800, 900]
}, index=['보행노인', '보행어린이', '자전거', '스쿨존', '기타'])

df_pie.plot.pie(
    y='사고건수',
    figsize=(7, 7),
    autopct='%.1f%%',
    startangle=90,
    counterclock=False,
    legend=False,
    shadow=True,
    wedgeprops={'width': 0.6, 'edgecolor': 'white'},
    ylabel=''   # y축 레이블 제거
)

plt.title('사고 유형별 건수 비율', fontsize=14)
plt.tight_layout()
plt.show()
````

## 14. pandas 히트맵 스타일 (보너스)

matplotlib의 `pcolormesh` 대신, pandas의 `.style.background_gradient()`를 사용하면  
**DataFrame 자체에 색상을 입혀** 간편하게 히트맵 효과를 낼 수 있음

````python
# ──────────────────────────────────────────────
# pandas 스타일 히트맵 (DataFrame에 직접 색상 적용)
# ──────────────────────────────────────────────

# 샘플 데이터 생성
rng = np.random.default_rng(42)
data = rng.integers(50, 200, size=(5, 4))
df_heat = pd.DataFrame(
    data,
    index=['보행노인', '보행어린이', '자전거', '스쿨존', '기타'],
    columns=['2021', '2022', '2023', '2024']
)

# 스타일 적용
styled = (
    df_heat
    .style
    .format("{:.0f}")                                    # 정수 형식
    .background_gradient(cmap='Oranges', axis=None)      # 전체 기준 색상
    .set_caption("사고유형 x 연도별 사고건수 (샘플)")     # 테이블 제목
)

display(styled)

# 엑셀 파일로 저장도 가능
# styled.to_excel('heatmap_styled.xlsx')
````

## 15. 한글 폰트 간편 설정 — koreanize-matplotlib

위의 **섹션 2**에서 OS별로 폰트를 수동 설정했지만,  
**`koreanize-matplotlib`** 패키지를 사용하면 **import 한 줄**로 한글 문제를 해결할 수 있음

> PyPI: https://pypi.org/project/koreanize-matplotlib/  
> GitHub: https://github.com/ychoi-kr/koreanize-matplotlib

### 특징

| 항목 | 내용 |
|------|------|
| **설치** | `pip install koreanize-matplotlib` |
| **사용법** | `import koreanize_matplotlib` (한 줄이면 끝) |
| **사용 폰트** | 나눔고딕 (Nanum Gothic) — 패키지에 포함되어 있어 별도 설치 불필요 |
| **라이선스** | MIT |

### 기존 방식 vs koreanize-matplotlib 비교

```python
# ──── 기존 방식 (OS별 분기 필요) ────
import platform
current_os = platform.system()
if current_os == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif current_os == 'Darwin':
    plt.rc('font', family='AppleGothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

# ──── koreanize-matplotlib (한 줄) ────
import koreanize_matplotlib
```
