## 9. 데이터의 특정 부분 선택 (Selection)

DataFrame에서 원하는 **행과 열을 골라내는 방법**

### 핵심 선택 도구 비교

| 방법 | 기준 | 예시 |
|------|------|------|
| `df['열이름']` | 열 이름 | `mydata['사고건수']` |
| `df[['열1','열2']]` | 여러 열 이름 | `mydata[['사고건수','사상자수']]` |
| `df.iloc[행, 열]` | **위치(숫자)** 기준 | `mydata.iloc[0, 2]` → 1행 3열 |
| `df.loc[행, 열]` | **이름(레이블)** 기준 | `mydata.loc[0, '사고건수']` |

````python
# ──────────────────────────────────────────────# 열(변수) 선택하기# ──────────────────────────────────────────────# 단일 변수 선택 → Series 반환print("=== 단일 변수 선택 (Series) ===")display(mydata['사상자수'].head(3))# 여러 변수 선택 → DataFrame 반환 (대괄호 2중!)print("\n=== 여러 변수 선택 (DataFrame) ===")display(mydata[['사고건수', '사상자수', '사망자수']].head(3))# filter 함수로 선택print("\n=== filter 함수로 선택 ===")display(mydata.filter(items=['사고건수', '사상자수']).head(3))
````

````python
# ──────────────────────────────────────────────# 행/열 선택: iloc (위치 기반) vs loc (이름 기반)# ──────────────────────────────────────────────# iloc: 정수 인덱스(위치)로 선택 — i는 integer의 iprint("=== iloc: 위치 기반 선택 ===")print("첫 번째 행, 모든 열:")display(mydata.iloc[0, :])     # 0번 행, 모든(:) 열print("\n처음 3행, 처음 4열:")display(mydata.iloc[:3, :4])   # 0~2행, 0~3열 (슬라이싱)# loc: 레이블(이름)로 선택print("\n=== loc: 이름 기반 선택 ===")print("0번 인덱스 행의 '사고건수':")print(mydata.loc[0, '사고건수'])
````

## 10. 데이터 전처리 (Cleaning)실제 데이터에는 **빈 값(결측치)**, **잘못된 자료형**, **이상한 값** 등이 존재할 수 있음  분석 전에 이러한 문제를 해결하는 과정 → **전처리(cleaning)**### 10-1. 결측치(Missing Values) 처리

````python
# ──────────────────────────────────────────────# 결측치 확인# ──────────────────────────────────────────────# 변수별 결측치 개수 확인print("=== 변수별 결측치 개수 ===")print(mydata.isnull().sum())# → 이 데이터는 결측치 없음. 실제 데이터에서는 자주 발생
````

````python
# ──────────────────────────────────────────────# 결측치 제거 / 채우기 (실습용 예시)# ──────────────────────────────────────────────# 결측치가 있는 모든 행 제거# mydata_clean = mydata.dropna()# 특정 변수에 결측치가 있는 행만 제거# mydata_clean = mydata.dropna(subset=['사고건수', '사상자수'])# 결측치를 특정 값으로 채우기# mydata['사망자수'] = mydata['사망자수'].fillna(0)         # 0으로 채우기# mydata['사고건수'] = mydata['사고건수'].fillna(mydata['사고건수'].mean())  # 평균으로 채우기
````

### 10-2. 자료형 변환 및 값 변환

````python
# ──────────────────────────────────────────────# 자료형 변환# ──────────────────────────────────────────────# 현재 자료형 확인print("=== 변수별 자료형 ===")print(mydata.dtypes)# 정수 → 실수로 변환mydata['사고건수'] = mydata['사고건수'].astype(float)# 숫자 → 문자열로 변환mydata['사고연도_str'] = mydata['사고연도'].astype(str)# 변환 결과 확인print(f"\n사고건수 타입: {mydata['사고건수'].dtype}")       # float64print(f"사고연도_str 타입: {mydata['사고연도_str'].dtype}")   # object (문자열)
````

````python
# ──────────────────────────────────────────────# 값 변환 (replace)# ──────────────────────────────────────────────# 특정 값을 다른 값으로 대체 (예시)# mydata['사고건수'] = mydata['사고건수'].replace(0, None)   # 0을 결측치로# mydata = mydata.replace({'보행어린이': 'Child Pedestrian', '보행노인': 'Elderly Pedestrian'})# 💡 임시 변수 정리: 사고건수를 다시 정수형으로 복원mydata['사고건수'] = mydata['사고건수'].astype(int)mydata.drop('사고연도_str', axis=1, inplace=True)
````

## 11. 조건으로 데이터 선택하기 (Filtering)

특정 조건을 만족하는 행만 골라내는 것 → **필터링(filtering)**  
엑셀의 "필터" 기능과 동일

### 비교 연산자

| 연산자 | 의미 | 예시 |
|--------|------|------|
| `==` | 같다 | `mydata['사고유형구분'] == '자전거'` |
| `!=` | 다르다 | `mydata['사고유형구분'] != '자전거'` |
| `>`, `>=` | 크다, 이상 | `mydata['사고건수'] >= 5` |
| `<`, `<=` | 작다, 이하 | `mydata['사고건수'] < 3` |
| `&` | 그리고 (AND) | 조건1 **&** 조건2 |
| `\|` | 또는 (OR) | 조건1 **\|** 조건2 |

````python
# ──────────────────────────────────────────────# 조건부 필터링# ──────────────────────────────────────────────# 1) 단일 조건: 사고건수 5건 이상filtered = mydata[mydata['사고건수'] >= 5]print(f"사고건수 5건 이상: {len(filtered)}행")# 2) AND 조건: 사고건수 5~10건filtered_and = mydata[(mydata['사고건수'] >= 5) & (mydata['사고건수'] <= 10)]print(f"사고건수 5~10건: {len(filtered_and)}행")# 3) OR 조건: 사고건수 10건 이상 또는 사망자수 1명 이상filtered_or = mydata[(mydata['사고건수'] >= 10) | (mydata['사망자수'] >= 1)]print(f"사고건수 10건+ 또는 사망자 1명+: {len(filtered_or)}행")# 4) 문자열 조건: 특정 사고유형만 선택filtered_type = mydata[mydata['사고유형구분'] == '자전거']print(f"자전거 사고만: {len(filtered_type)}행")# 5) isin(): 여러 값 중 하나에 해당하는 행 선택filtered_multi = mydata[mydata['사고유형구분'].isin(['자전거', '보행어린이'])]print(f"자전거+보행어린이: {len(filtered_multi)}행")
````

````python
# ──────────────────────────────────────────────# query() 메서드: 더 읽기 쉬운 필터링# ──────────────────────────────────────────────# query()는 조건을 문자열로 작성 — SQL 쿼리와 유사하여 직관적filtered_q1 = mydata.query('사고건수 >= 5')filtered_q2 = mydata.query('사고건수 >= 5 and 사고건수 <= 10')filtered_q3 = mydata.query('사고유형구분 == "자전거"')filtered_q4 = mydata.query('사고유형구분 in ["자전거", "보행어린이"]')print(f"query 결과 - 자전거+보행어린이: {len(filtered_q4)}행")display(filtered_q4.head(3))
````

## 12. 데이터 정렬 (Sorting)**특정 변수 기준으로 오름차순/내림차순 정렬** 가능

````python
# ──────────────────────────────────────────────# 변수 기준 정렬# ──────────────────────────────────────────────# 사고건수 기준 오름차순 (적은 것부터)sorted_asc = mydata.sort_values('사고건수')print("=== 사고건수 오름차순 (상위 3행) ===")display(sorted_asc.head(3))# 사고건수 기준 내림차순 (많은 것부터)sorted_desc = mydata.sort_values('사고건수', ascending=False)print("\n=== 사고건수 내림차순 (상위 3행) ===")display(sorted_desc[['사고지역위치명', '사고건수', '사상자수', '사고유형구분']].head(3))# nlargest / nsmallest: 상위/하위 N개만 빠르게 추출print("\n=== 사고건수 상위 5개 지역 ===")display(mydata.nlargest(5, '사고건수')[['사고지역위치명', '사고건수', '사상자수', '사고유형구분']])
````

## 13. 새로운 변수(열) 만들기기존 변수를 조합하거나 계산하여 **새로운 변수 추가** 가능  → **파생변수(derived variable)** 생성

````python
# ──────────────────────────────────────────────# 새 변수 생성# ──────────────────────────────────────────────# 1) 산술 연산으로 비율 변수 생성mydata['사고당_사상자수'] = mydata['사상자수'] / mydata['사고건수']# 2) 여러 변수 조합mydata['총_부상자'] = mydata['중상자수'] + mydata['경상자수'] + mydata['부상신고자수']# 3) 문자열 조합으로 새 변수mydata['사고연도_표시'] = mydata['사고연도'].astype(str) + '년'# 결과 확인print("=== 새로 만든 변수 확인 ===")display(mydata[['사고지역위치명', '사고건수', '사상자수', '사고당_사상자수', '총_부상자', '사고연도_표시']].head(5))
````

## 14. 집단 분류와 통계 분석 (GroupBy)**그룹별로 묶어서 통계를 계산**하는 것 → 데이터 분석의 핵심  엑셀의 "부분합" 또는 SQL의 `GROUP BY`와 동일한 기능```원본 데이터 → groupby('기준변수') → 집계함수(sum, mean 등) → 결과 테이블```

````python
# ──────────────────────────────────────────────# 기본 그룹화: 사고유형별 통계# ──────────────────────────────────────────────by_type = mydata.groupby('사고유형구분')# 사고유형별 주요 통계print("=== 사고유형별 사고건수 합계 ===")display(by_type['사고건수'].sum())print("\n=== 사고유형별 사고건수 평균 ===")display(by_type['사고건수'].mean().round(2))print("\n=== 사고유형별 빈도(데이터 행 수) ===")display(by_type['사고건수'].count())
````

````python
# ──────────────────────────────────────────────# 다중 그룹화 + 여러 집계함수 동시 적용# ──────────────────────────────────────────────# 사고유형 × 사고연도별 사고건수 합계와 평균result = mydata.groupby(['사고유형구분', '사고연도']).agg(    사고건수_합계=('사고건수', 'sum'),    사고건수_평균=('사고건수', 'mean'),    사상자수_합계=('사상자수', 'sum')).round(2)print("=== 사고유형 × 연도별 통계 (상위 10행) ===")display(result.head(10))
````

## 15. 피벗 테이블 (Pivot Table)**피벗 테이블**: GroupBy의 확장 버전으로, 행과 열을 **교차시켜** 요약  엑셀의 "피벗 테이블"과 동일한 개념```pivot_table(    index="행이 될 변수",    columns="열이 될 변수",    values="계산할 변수",    aggfunc="집계함수")```

````python
# ──────────────────────────────────────────────# 기본 피벗 테이블# ──────────────────────────────────────────────# 사고유형(행) × 사고연도(열)별 사고건수 합계pivot1 = mydata.pivot_table(    index="사고유형구분",       # 행: 사고유형    columns="사고연도",         # 열: 사고연도    values="사고건수",          # 계산 대상: 사고건수    aggfunc="sum",             # 집계 함수: 합계    fill_value=0               # 결측치를 0으로)print("=== 사고유형 × 연도별 사고건수 합계 ===")display(pivot1)
````

````python
# ──────────────────────────────────────────────# 고급 피벗 테이블: 여러 집계 함수 + 합계 행/열# ──────────────────────────────────────────────# 여러 집계 함수 적용pivot2 = mydata.pivot_table(    index="사고유형구분",    values=["사고건수", "사상자수"],    aggfunc={"사고건수": ["sum", "mean"], "사상자수": ["sum", "max"]},    fill_value=0).round(2)print("=== 여러 집계 함수 적용 ===")display(pivot2)# 합계 행/열 추가 (margins)pivot3 = mydata.pivot_table(    index="사고유형구분",    columns="사고연도",    values="사고건수",    aggfunc="sum",    margins=True,          # 합계 행/열 추가    margins_name="총계",    # 합계 이름    fill_value=0)print("\n=== 합계 포함 피벗 테이블 (최근 3년) ===")display(pivot3[[2022, 2023, 2024, '총계']])
````

````python
# ──────────────────────────────────────────────# apply + lambda로 변수 변환# ──────────────────────────────────────────────# 1) 단일 변수에 함수 적용: 비율(%) 계산data_pct = mydata[["사고건수", "사상자수", "중상자수"]].apply(    lambda x: x / x.sum() * 100   # 각 열의 합계 대비 비율).round(4)print("=== 변수별 비율(%) 상위 3행 ===")display(data_pct.head(3))
````

````python
# ──────────────────────────────────────────────# 조건부 변수 생성# ──────────────────────────────────────────────# 방법 1: def 함수 정의 후 applydef 심각도_평가(row):    """사고건수 기준 심각도 분류 함수"""    if row['사고건수'] >= 7:        return '높음'    elif row['사고건수'] >= 4:        return '보통'    else:        return '낮음'mydata['심각도'] = mydata.apply(심각도_평가, axis=1)  # axis=1: 행 단위로 적용# 방법 2: lambda로 간단하게 (2단계 분류)# mydata['심각도'] = mydata['사고건수'].apply(lambda x: '높음' if x >= 7 else '낮음')# 결과 확인print("=== 심각도별 빈도 ===")print(mydata['심각도'].value_counts())display(mydata[['사고지역위치명', '사고건수', '심각도']].head(5))
````

## 17. 데이터 합치기 (Combining)

여러 데이터를 하나로 합치는 것 → 매우 자주 사용되는 기능

### 합치기 방법 비교

| 함수 | 방향 | 기준 | 비유 |
|------|------|------|------|
| `pd.concat(axis=0)` | 위아래 (행) | 같은 변수명 | 테이블 아래에 붙이기 |
| `pd.concat(axis=1)` | 좌우 (열) | 같은 인덱스 | 테이블 옆에 붙이기 |
| `pd.merge()` | 좌우 (열) | **공통 변수 값** | SQL JOIN |

```
concat(axis=0):          concat(axis=1):          merge(on='키'):
┌──────┐                 ┌──────┬──────┐         ┌──────┬──────┐
│ df1  │                 │ df1  │ df3  │         │ df1  │ df4  │
├──────┤  → 행 합치기    └──────┴──────┘         └──────┴──────┘
│ df2  │                   → 열 합치기              → 키 기준 합치기
└──────┘
```

````python
# ──────────────────────────────────────────────# 데이터 합치기 실습용 샘플 생성# ──────────────────────────────────────────────data1 = pd.DataFrame({"변수1": [1, 2, 3], "변수2": [5, 3, 1]})data2 = pd.DataFrame({"변수1": [21, 43], "변수2": [54, 12]})data3 = pd.DataFrame({"변수3": [224, 453, 648]})data4 = pd.DataFrame({"변수1": [1, 2, 5], "변수3": [224, 453, 648]})# 1) 행 방향 합치기 (위아래로)combined_rows = pd.concat([data1, data2], axis=0, ignore_index=True)print("=== 행 방향 합치기 (concat axis=0) ===")display(combined_rows)# 2) 열 방향 합치기 (좌우로)combined_cols = pd.concat([data1, data3], axis=1)print("\n=== 열 방향 합치기 (concat axis=1) ===")display(combined_cols)
````

````python
# ──────────────────────────────────────────────# merge: 공통 변수(키) 기준으로 합치기 — SQL JOIN과 동일# ──────────────────────────────────────────────print("data1:")display(data1)print("\ndata4:")display(data4)# inner join: 양쪽 모두에 있는 키 값만 (교집합)merged_inner = pd.merge(data1, data4, on="변수1", how="inner")print("\n=== Inner Join (교집합) ===")display(merged_inner)# outer join: 양쪽 어디에든 있는 키 값 모두 (합집합)merged_outer = pd.merge(data1, data4, on="변수1", how="outer")print("\n=== Outer Join (합집합) ===")display(merged_outer)# left join: 왼쪽 데이터의 키 값 기준merged_left = pd.merge(data1, data4, on="변수1", how="left")print("\n=== Left Join (왼쪽 기준) ===")display(merged_left)
````
