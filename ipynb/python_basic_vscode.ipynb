### 5-3. 리스트 컴프리헨션 (List Comprehension)리스트 각 요소에 **같은 변환을 일괄 적용**하여 새 리스트를 만드는 방법```python[변환식 for 요소 in 리스트]              # 기본[변환식 for 요소 in 리스트 if 조건]      # 조건 포함```

````python
# ──────────────────────────────────────────────# 리스트 컴프리헨션# ──────────────────────────────────────────────lst_num = [23, 52, 98, 31]# 각 요소를 2로 나눈 나머지remainders = [i % 2 for i in lst_num]print(f"나머지: {remainders}")# 조건 포함: 글자 수가 1이 아닌 멤버만 추출lst_string = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]filtered = [name for name in lst_string if len(name) != 1]print(f"1글자 제외: {filtered}")
````

## 13. 클래스(Class)**객체를 찍어내는 설계도(틀)**- 클래스(class): 설계도 (예: 붕어빵 틀)- 인스턴스(instance): 클래스로 만든 실제 객체 (예: 붕어빵 낱개)- 메서드(method): 클래스 안에 정의된 함수 (예: 굽기, 꺼내기)- 생성자(`__init__`): 객체 생성 시 자동 호출되는 초기화 메서드- 속성(attribute): 클래스/인스턴스가 가지는 변수```pythonclass 클래스이름:    def __init__(self, 매개변수):    # 생성자        self.속성 = 매개변수         # 속성 정의    def 메서드(self):                # 메서드 정의        return 결과```

````python
# ──────────────────────────────────────────────# 계산기 클래스# ──────────────────────────────────────────────class Calculation:    def __init__(self, n1, n2):   # 생성자: 두 숫자를 입력받아 속성으로 저장        self.n1 = n1              # 속성 1        self.n2 = n2              # 속성 2    def add(self):                # 메서드: 더하기        return self.n1 + self.n2    def sub(self):                # 메서드: 빼기        return self.n1 - self.n2    def mul(self):                # 메서드: 곱하기        return self.n1 * self.n2    def div(self):                # 메서드: 나누기        return self.n1 / self.n2# 객체 생성 (클래스는 대문자, 객체는 소문자 관례)cal = Calculation(8, 5)print(f"n1={cal.n1}, n2={cal.n2}")   # 속성 확인print(f"더하기: {cal.add()}")print(f"빼기: {cal.sub()}")print(f"곱하기: {cal.mul()}")print(f"나누기: {cal.div()}")
````

````python
# ──────────────────────────────────────────────# 실전 예제: BTS 멤버 클래스# ──────────────────────────────────────────────members = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]class MemberBTS:    def __init__(self, member, age):        self.member = member        self.age = age    def introduce(self):        if self.member in members:            return f"저는 BTS 멤버인 {self.member}이며, {self.age}살"        else:            return f"저는 BTS 멤버가 아닌 {self.member}이며, {self.age}살"m1 = MemberBTS("뷔", 30)m2 = MemberBTS("방시혁", 53)print(m1.introduce())print(m2.introduce())
````
