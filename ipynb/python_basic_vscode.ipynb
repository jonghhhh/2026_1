## 1. 파이썬 코드 작성 원리

### 핵심 구성 요소

| 구성 요소 | 설명 | 예시 |
|----------|------|------|
| **객체(object)** | 데이터를 담는 그릇. 파이썬에서 **모든 것이 객체** (숫자, 문자, 리스트, 함수 등) | `text = "안녕"` |
| **함수(function)** | 객체에 **변화를 가하는 동작** | `.split()`, `print()` |
| **인자(argument)** | 함수에 전달하는 **조건/입력값** | `split("7")` 의 `"7"` |

### 코딩 기본 패턴

```python
객체.함수(인자)
```
→ 특정 자료(객체)에 대해 특정 조건(인자)으로 변화를 가함(함수)  
→ 결과를 다른 객체에 **할당(=)** 가능  
→ 파이썬 **내장 함수**(print, len, type 등)는 객체 선행 없이 단독 사용 가능

````python
# ──────────────────────────────────────────────
# 객체 생성과 함수 사용 기본 패턴
# ──────────────────────────────────────────────

# 1) 객체 생성 (자료의 할당)
text = "BTS는 대한민국 7인조 보이 그룹이다."
text
````

````python
# 2) <객체.함수(인자)> 방식으로 코딩
text.split("7")    # text를 "7"을 기준으로 분리 → 리스트 반환
````

````python
# 3) 결과를 새로운 객체에 할당
text_1 = text.split(" ")    # 공백 기준 분리 후 text_1에 할당
print(text_1)
print(type(text_1))          # list 타입
````

````python
# 4) 파이썬 내장 함수 — 객체 없이 단독 사용
print(text)          # 출력
print(type(text))    # 자료형 확인 → str
print(len(text))     # 길이(글자 수)
print(round(3.5678, 2))  # 반올림 (소수점 2자리) → 3.57
````

## 2. 문자열(String)

문자, 단어 등으로 구성된 자료형
- 큰따옴표(`"..."`) 또는 작은따옴표(`'...'`)로 감싸서 생성
- 여러 줄 문자열: 따옴표 3개(`"""..."""` 또는 `'''...'''`)로 감싸기

````python
# ──────────────────────────────────────────────
# 문자열 생성
# ──────────────────────────────────────────────

s = "안녕하세요"                    # 큰따옴표
s_1 = 'BTS가 "안녕하세요"라고 인사했다'  # 작은따옴표 (안에 큰따옴표 포함 가능)

# 여러 줄 문자열
s_long = '''어느 날 세상이 멈췄어
아무런 예고도 하나 없이
봄은 기다림을 몰라서
눈치 없이 와버렸어'''

print(s)
print(s_1)
print(s_long)    # \n(줄바꿈 문자)이 자동 포함됨
````

### 2-1. 문자열 연산 및 길이

````python
# ──────────────────────────────────────────────
# 문자열 연산
# ──────────────────────────────────────────────

a = "안녕하세요 "
b = "BTS입니다"

print(a + b)     # 문자열 연결 (concatenation)
print(a * 3)     # 문자열 반복
print(len(a))    # 길이(글자 수) — 공백도 1글자로 카운트
````

### 2-2. 문자열 관련 함수

````python
# ──────────────────────────────────────────────
# 문자열 검색/확인 함수
# ──────────────────────────────────────────────

s = "안녕하세요"

print(s.startswith("안녕"))   # "안녕"으로 시작하는지 → True
print(s.endswith("세요"))     # "세요"로 끝나는지 → True
print(s.find("하"))          # "하"가 처음 나오는 위치(인덱스) → 2
print(s.count("하"))         # "하"의 개수 → 1
````

````python
# ──────────────────────────────────────────────
# 문자열 변환 함수
# ──────────────────────────────────────────────

print("Hello World".lower())     # 소문자로 변환
print("Hello World".upper())     # 대문자로 변환
print("  안녕하세요  ".strip())    # 양쪽 공백 제거
print("안녕하세요".replace("안녕", "반갑"))  # 문자열 대체
print("/".join("ABC"))           # 각 문자 사이에 "/" 삽입 → A/B/C
````

### 2-3. 문자열 포매팅 (Formatting)

문자열 안에 **변수 값을 삽입**하는 3가지 방법

| 방법 | 문법 | 특징 |
|------|------|------|
| **f-string** | `f"텍스트 {변수}"` | ✅ 가장 권장 (Python 3.6+) |
| format() | `"텍스트 {}".format(값)` | 범용적 |
| % 연산자 | `"텍스트 %s" % 값` | 구식, 참고용 |

````python
# ──────────────────────────────────────────────
# 문자열 포매팅
# ──────────────────────────────────────────────
name = "BTS"
age = 13    # 데뷔(2013년) 이후 년수 (2026년 기준)

# ✅ f-string (권장)
print(f"{name}는 데뷔 {age}년차 그룹")

# format() 메서드
print("{} 더하기 {} 은 {}".format(3, 4, 7))

# % 연산자 (구식)
print("%s 더하기 %d 은 %d" % ("3", 4, 7))
````

### 2-4. 인덱싱(Indexing)과 슬라이싱(Slicing)

문자열의 **특정 위치 글자를 가리키거나(인덱싱)** **범위를 잘라내기(슬라이싱)**

> **"파이썬은 0부터 숫자를 센다"**

```
 문자열:  안  녕  하  세  요
 인덱스:  0   1   2   3   4
 음수:   -5  -4  -3  -2  -1
```

````python
# ──────────────────────────────────────────────
# 인덱싱 & 슬라이싱
# ──────────────────────────────────────────────

s = "안녕하세요"

# 인덱싱: 특정 위치의 글자 1개
print(s[0])     # 첫 번째 글자 → "안"
print(s[-1])    # 마지막 글자 → "요"

# 슬라이싱: [시작:끝] — 시작 이상, 끝 미만
print(s[0:3])   # 0~2번 → "안녕하"
print(s[2:])    # 2번부터 끝까지 → "하세요"
print(s[:3])    # 처음부터 2번까지 → "안녕하"
print(s[:])     # 전체 복사
````

## 3. 숫자(Numeric)

숫자 형태의 자료형
- **정수형(int)**: 소수점이 없는 숫자 (예: 5, -3, 0)
- **실수형(float)**: 소수점이 있는 숫자 (예: 2.5, -0.1)

````python
# ──────────────────────────────────────────────
# 숫자 자료형 생성 및 변환
# ──────────────────────────────────────────────

i = int("5")       # 문자열 "5" → 정수 5
f = float("2.5")   # 문자열 "2.5" → 실수 2.5

print(f"정수: {i}, 타입: {type(i)}")
print(f"실수: {f}, 타입: {type(f)}")
````

### 3-1. 산술 연산자

````python
# ──────────────────────────────────────────────
# 산술 연산자 한눈에 보기
# ──────────────────────────────────────────────

print(f"5 + 3 = {5 + 3}")     # 더하기
print(f"5 - 3 = {5 - 3}")     # 빼기
print(f"5 * 3 = {5 * 3}")     # 곱하기
print(f"5 / 3 = {5 / 3:.2f}") # 나누기 (결과는 항상 float)
print(f"5 ** 2 = {5 ** 2}")    # 거듭제곱 (5의 2제곱)
print(f"7 // 3 = {7 // 3}")   # 몫
print(f"7 % 3 = {7 % 3}")     # 나머지

# 복합 할당 연산자
x = 10
x += 3    # x = x + 3 → 13
x -= 1    # x = x - 1 → 12
print(f"x = {x}")
````

## 4. 불(Bool) — 논리 자료형

**참(True)**과 **거짓(False)** 2가지 값만 가지는 자료형  
조건문(if)과 함께 자주 사용

````python
# ──────────────────────────────────────────────
# 불 자료형과 비교/논리 연산
# ──────────────────────────────────────────────

print(2 > 1)           # True
print(type(2 > 1))     # <class 'bool'>

# 논리 연산자: and(둘 다 참), or(하나라도 참), not(반대)
print((2 > 1) and (0 > 1))   # True and False → False
print((2 > 1) or (0 > 1))    # True or False → True
print(not (2 > 1))            # not True → False

# in 연산자: 포함 여부 확인
a = "안녕하세요. BTS입니다"
print("BTS" in a)              # True
````

## 5. 리스트(List)

여러 개의 요소를 **순서대로** 담는 자료형
- **대괄호(`[ ]`)**로 감싸고, 각 요소는 **쉼표(`,`)** 로 구분
- 문자열, 숫자, 불, 심지어 리스트까지 혼합 가능
- **수정, 추가, 삭제 가능** (가변, mutable)

````python
# ──────────────────────────────────────────────
# 리스트 생성
# ──────────────────────────────────────────────

lst_numeric = [23, 52, 98, 31]                                  # 숫자 리스트
lst_string = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]   # 문자 리스트
lst_mix = ["RM", 23, True, [1, 2]]                               # 혼합 리스트
lst_empty = []                                                    # 빈 리스트

print(f"숫자 리스트: {lst_numeric}")
print(f"문자 리스트: {lst_string}")
````

### 5-1. 리스트 연산, 인덱싱, 슬라이싱

````python
# ──────────────────────────────────────────────
# 리스트 연산
# ──────────────────────────────────────────────

lst_string = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]

print(lst_string + ["아미"])    # 연결
print(lst_string * 2)           # 반복
print(len(lst_string))          # 길이(요소 수) → 7

# 인덱싱
print(lst_string[0])     # 첫 번째 → "RM"
print(lst_string[-1])    # 마지막 → "정국"

# 슬라이싱
print(lst_string[1:4])   # 2~4번째 → ["진", "슈가", "제이홉"]
print(lst_string[3:])    # 4번째부터 끝까지
````

### 5-2. 리스트 수정, 삭제, 관련 함수

````python
# ──────────────────────────────────────────────
# 리스트 수정 및 삭제
# ──────────────────────────────────────────────

lst = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]

# 수정: 인덱스로 직접 변경
lst[0] = "남준"
print(f"수정 후: {lst}")

# 삭제: del 또는 remove
del lst[0]                # 인덱스로 삭제
print(f"del 후: {lst}")

lst.remove("뷔")           # 값으로 삭제
print(f"remove 후: {lst}")
````

````python
# ──────────────────────────────────────────────
# 리스트 주요 함수
# ──────────────────────────────────────────────

lst = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]

lst.append("방시혁")       # 맨 뒤에 추가
lst.insert(0, "아미")      # 0번 위치에 삽입
print(f"append+insert: {lst}")

lst.remove("아미")          # 특정 요소 제거
print(f"remove 후: {lst}")

lst.extend(["아미1", "아미2"])  # 리스트 확장 (개별 요소로 추가)
print(f"extend 후: {lst}")

print(f"index('진'): {lst.index('진')}")   # 위치 반환
print(f"count('진'): {lst.count('진')}")   # 개수 반환

lst_num = [23, 52, 98, 31]
print(f"sum: {sum(lst_num)}, min: {min(lst_num)}, max: {max(lst_num)}")
print(f"평균: {sum(lst_num)/len(lst_num)}")

lst_num.sort()              # 오름차순 정렬 (원본 변경)
print(f"정렬 후: {lst_num}")
````

### 5-3. 리스트 컴프리헨션 (List Comprehension)

리스트 각 요소에 **같은 변환을 일괄 적용**하여 새 리스트를 만드는 방법

```python
[변환식 for 요소 in 리스트]              # 기본
[변환식 for 요소 in 리스트 if 조건]      # 조건 포함
```

````python
# ──────────────────────────────────────────────
# 리스트 컴프리헨션
# ──────────────────────────────────────────────

lst_num = [23, 52, 98, 31]

# 각 요소를 2로 나눈 나머지
remainders = [i % 2 for i in lst_num]
print(f"나머지: {remainders}")

# 조건 포함: 글자 수가 1이 아닌 멤버만 추출
lst_string = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]
filtered = [name for name in lst_string if len(name) != 1]
print(f"1글자 제외: {filtered}")
````

## 6. 사전(Dictionary)

**key와 value의 쌍(대응관계)**으로 이루어진 자료형
- **중괄호(`{ }`)**로 감싸고, `key: value` 형태
- key는 고유해야 하며, value는 중복 가능
- **순서 보장** (Python 3.7+)

````python
# ──────────────────────────────────────────────
# 사전 생성 및 기본 조작
# ──────────────────────────────────────────────

dct = {"한국": "서울", "미국": "워싱턴", "중국": "베이징"}

# 요소 출력
print(f"keys: {list(dct.keys())}")
print(f"values: {list(dct.values())}")
print(f"items: {list(dct.items())}")

# 인덱싱 (key로 value 접근)
print(f"한국의 수도: {dct['한국']}")
print(f"일본의 수도: {dct.get('일본', '데이터 없음')}")  # 없는 key → 기본값 반환

# 요소 추가 / 삭제
dct["영국"] = "런던"      # 추가
print(f"추가 후: {dct}")
del dct["영국"]           # 삭제
print(f"삭제 후: {dct}")
````

## 9. if문 (조건문)

조건을 판단하여 **해당 조건에 맞는 코드만 실행**
- `if` → `elif` → `else` 순서로 조건 분기
- **콜론(`:`)과 들여쓰기(4칸 공백)** 필수

```python
if 조건1:
    실행문1          # 조건1이 True일 때
elif 조건2:
    실행문2          # 조건1은 False이고 조건2가 True일 때
else:
    실행문3          # 위 모든 조건이 False일 때
```

````python
# ──────────────────────────────────────────────
# 기본 if-elif-else
# ──────────────────────────────────────────────

members = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]

x = "진"
if x in members:
    print(f"{x}은/는 BTS 멤버")
else:
    print(f"{x}은/는 BTS 멤버가 아님")
````

````python
# ──────────────────────────────────────────────
# 비교 연산자 활용: <, >, >=, <=, ==, !=
# ──────────────────────────────────────────────

x = 3
if x > 5:
    print(f"{x}은 5보다 큼")
elif x < 0:
    print(f"{x}은 음수")
else:
    print(f"{x}은 0과 5 사이에 위치")
````

````python
# ──────────────────────────────────────────────
# 논리 연산자: and, or, not / pass 사용
# ──────────────────────────────────────────────

members = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]
p1, p2 = "지민", "슈가"

if (p1 in members) and (p2 in members):
    print(f"{p1}과 {p2}은 모두 BTS 멤버")
elif (p1 in members) or (p2 in members):
    print(f"{p1}과 {p2} 중 적어도 한 명은 BTS 멤버")
else:
    print(f"{p1}과 {p2}은 모두 BTS 멤버가 아님")

# pass: 조건에 해당하지만 아무것도 실행하지 않고 넘어가기
p = "방시혁"
if p in members:
    print(f"{p}은 BTS 멤버")
elif p == "방시혁":
    pass    # 아무 동작 없이 넘어감
else:
    print(f"{p}은 BTS 멤버가 아님")
````

## 10. while문 (반복문)

`while` 뒤 조건이 **True인 동안 계속 반복** 실행

```python
while 조건:
    실행문        # 조건이 True인 동안 반복
```

> ⚠️ 종료 조건이 없으면 **무한 루프**에 빠지므로 주의

````python
# ──────────────────────────────────────────────
# while문 기본
# ──────────────────────────────────────────────

# 구구단 2단
x = 1
while x <= 9:
    print(f"2 × {x} = {2 * x}")
    x += 1    # x를 1씩 증가 → x가 10이 되면 조건 불만족으로 종료
````

````python
# ──────────────────────────────────────────────
# while + break: 반복 도중 강제 종료
# ──────────────────────────────────────────────

members = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]

x = 1
while True:                  # 무한 반복
    print(members[:x])
    x += 1
    if x > len(members):     # 멤버 수를 초과하면
        break                # 반복 종료
````

## 11. for문 (반복문)

```python
for 변수 in 반복가능객체:
    실행문
```

> while문과의 차이: while은 **조건 기반**, for는 **요소 순회 기반**

````python
# ──────────────────────────────────────────────
# for문 기본
# ──────────────────────────────────────────────

members = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]

for person in members:
    print(f"{person}는 BTS 멤버")
````

````python
# ──────────────────────────────────────────────
# for + if 조합: 조건부 반복
# ──────────────────────────────────────────────

시험 = [95, 55, 71, 83, 79, 51, 90]
평가 = []

for 점수 in 시험:
    if 점수 >= 80:
        평가.append("합격")
    else:
        평가.append("불합격")

print(f"점수: {시험}")
print(f"평가: {평가}")
````

### 11-1. range() 함수

for문과 함께 자주 사용

````python
# ──────────────────────────────────────────────
# range()와 for문
# ──────────────────────────────────────────────

# range(0, 4) → 0, 1, 2, 3
members = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]
for i in range(0, 4):
    print(members[:i+1])
````

````python
# ──────────────────────────────────────────────
# 중첩 for문: 구구단 만들기
# ──────────────────────────────────────────────

for i in range(2, 10):       # 2단 ~ 9단
    for j in range(1, 10):   # ×1 ~ ×9
        print(f"{i} × {j} = {i*j}")
    print()                  # 단 사이 빈 줄
````

### 11-2. 리스트 컴프리헨션 (for문의 축약)

````python
# ──────────────────────────────────────────────
# 리스트 컴프리헨션 + 조건
# ──────────────────────────────────────────────

numbers = list(range(1, 16))   # [1, 2, ..., 15]

# 2단 만들기 (9 미만만)
result = [f"{i} × 2 = {i*2}" for i in numbers if i < 10]
print(result)

# if-else 포함 (else는 for 앞에 위치)
result2 = [f"{i} × 2 = {i*2}" if i < 10 else "범위초과" for i in numbers]
print(result2)
````

## 12. 함수(Function)

입력값을 받아 **특정 작업을 수행**한 후 결과를 반환하는 코드 묶음
- **반복되는 코드**를 하나로 묶어 재사용
- 프로그램 흐름 파악과 오류 탐지에 유용

```python
def 함수이름(매개변수):

```

| 용어 | 설명 | 예시 |
|------|------|------|
| **매개변수(parameter)** | 함수 정의 시 입력을 받는 변수 | `def add(x, y):` 의 x, y |

### 12-1. 다양한 함수 형태

````python
# ──────────────────────────────────────────────
# 키워드 인수 / 초깃값 설정
# ──────────────────────────────────────────────

def 인사(누가, 누구에게, 어떻게="반갑게"):
    """인사 문장 생성 함수 (어떻게는 기본값 '반갑게')"""
    return f"{누가}가 {누구에게}에게 {어떻게} '안녕하세요'라고 인사한다"

print(인사("유재석", "BTS"))                    # 기본값 사용
print(인사("유재석", "BTS", "다정하게"))          # 기본값 대체
print(인사(누구에게="BTS", 누가="유재석"))        # 키워드로 순서 무관
````

````python
# ──────────────────────────────────────────────
# *args: 가변 인수 (개수 제한 없는 입력)
# ──────────────────────────────────────────────

def multiple_add(*args):
    """입력된 모든 숫자의 합을 반환. *args는 입력값을 튜플로 묶어줌"""
    return sum(args)

print(multiple_add(2, 3, 4))       # 9
print(multiple_add(1, 2, 3, 4, 5)) # 15
````

````python
# ──────────────────────────────────────────────
# **kwargs: 키워드 가변 인수 (딕셔너리로 받기)
# ──────────────────────────────────────────────

def country_capital(**kwargs):
    """국가=수도 형태의 키워드 인수를 받아 출력"""
    result = []
    for key, value in kwargs.items():
        result.append(f"{key}의 수도는 {value}")
    return result

print(country_capital(한국="서울", 중국="베이징", 미국="워싱턴"))

# 💡 함수 인자 순서 규칙: func(일반, *args, **kwargs)
````

## 13. 클래스(Class)

**객체를 찍어내는 설계도(틀)**
- 클래스(class): 설계도 (예: 붕어빵 틀)
- 인스턴스(instance): 클래스로 만든 실제 객체 (예: 붕어빵 낱개)
- 메서드(method): 클래스 안에 정의된 함수 (예: 굽기, 꺼내기)
- 생성자(`__init__`): 객체 생성 시 자동 호출되는 초기화 메서드
- 속성(attribute): 클래스/인스턴스가 가지는 변수

```python
class 클래스이름:
    def __init__(self, 매개변수):    # 생성자
        self.속성 = 매개변수         # 속성 정의
    def 메서드(self):                # 메서드 정의
        return 결과
```

````python
# ──────────────────────────────────────────────
# 계산기 클래스
# ──────────────────────────────────────────────

class Calculation:
    def __init__(self, n1, n2):   # 생성자: 두 숫자를 입력받아 속성으로 저장
        self.n1 = n1              # 속성 1
        self.n2 = n2              # 속성 2

    def add(self):                # 메서드: 더하기
        return self.n1 + self.n2

    def sub(self):                # 메서드: 빼기
        return self.n1 - self.n2

    def mul(self):                # 메서드: 곱하기
        return self.n1 * self.n2

    def div(self):                # 메서드: 나누기
        return self.n1 / self.n2

# 객체 생성 (클래스는 대문자, 객체는 소문자 관례)
cal = Calculation(8, 5)

print(f"n1={cal.n1}, n2={cal.n2}")   # 속성 확인
print(f"더하기: {cal.add()}")
print(f"빼기: {cal.sub()}")
print(f"곱하기: {cal.mul()}")
print(f"나누기: {cal.div()}")
````

````python
# ──────────────────────────────────────────────
# 실전 예제: BTS 멤버 클래스
# ──────────────────────────────────────────────

members = ["RM", "진", "슈가", "제이홉", "지민", "뷔", "정국"]

class MemberBTS:
    def __init__(self, member, age):
        self.member = member
        self.age = age

    def introduce(self):
        if self.member in members:
            return f"저는 BTS 멤버인 {self.member}이며, {self.age}살"
        else:
            return f"저는 BTS 멤버가 아닌 {self.member}이며, {self.age}살"

m1 = MemberBTS("뷔", 30)
m2 = MemberBTS("방시혁", 53)

print(m1.introduce())
print(m2.introduce())
````

## 16. 파일 쓰기와 읽기 (txt 파일)

`open()` 함수로 파일을 열고, `with` 문을 사용하면 **자동으로 파일을 닫아줌**

| 모드 | 설명 |
|------|------|
| `"w"` | 쓰기 (기존 내용 덮어쓰기) |
| `"r"` | 읽기 |
| `"a"` | 추가 (기존 내용 뒤에 이어쓰기) |

````python
# ──────────────────────────────────────────────
# 파일 쓰기와 읽기 (VSCode 로컬 환경)
# ──────────────────────────────────────────────

# 파일 쓰기
with open("test.txt", "w", encoding="utf-8") as f:
    f.write("RM, 진, 슈가, 제이홉, 지민, 뷔, 정국")
print("✅ test.txt 파일 저장 완료")

# 파일 읽기
with open("test.txt", "r", encoding="utf-8") as f:
    mytext = f.read()
print(f"파일 내용: {mytext}")

# 💡 VSCode 환경: 노트북과 같은 폴더에 파일이 생성됨
````

## 📋 파이썬 기초 요약 (Quick Reference)

| 카테고리 | 핵심 문법 | 설명 |
|---------|----------|------|
| **자료형** | `str`, `int`, `float`, `bool` | 기본 자료형 |
| **컬렉션** | `list[]`, `dict{}`, `tuple()`, `set{}` | 여러 값을 담는 자료구조 |
| **조건문** | `if`, `elif`, `else` | 조건 분기 |
| **반복문** | `for ... in`, `while` | 반복 실행 |
| **함수** | `def 함수명(매개변수): return 값` | 코드 재사용 |
| **lambda** | `lambda x: 표현식` | 한 줄 함수 |
| **클래스** | `class 이름: def __init__(self):` | 객체 설계도 |
| **모듈** | `import 모듈명` / `from 모듈 import 함수` | 코드 재사용 (파일 단위) |
| **패키지** | `import 패키지.모듈` | 모듈의 디렉터리 구조 |
| **파일** | `open("파일", "r/w")` | 파일 읽기/쓰기 |
