## 0. OpenCV란?

**OpenCV (Open Source Computer Vision Library)**
2000년 인텔에 의해 출시된 **컴퓨터 비전 및 기계 학습** 오픈소스 라이브러리

### 주요 특징

| 항목 | 내용 |
|------|------|
| **탄생** | 2000년 인텔에서 시작, 현재 OpenCV.org에서 관리 |
| **언어** | C++로 작성, **Python 바인딩** 제공 (cv2 모듈) |
| **라이선스** | Apache 2.0 (상업적 사용 가능) |
| **최신 버전** | 4.13.0 (2026년 2월 기준) |

### 핵심 기능

| 기능 | 설명 | 함수 예시 |
|------|------|----------|
| 이미지 읽기/쓰기 | 다양한 형식의 이미지 입출력 | `cv2.imread()`, `cv2.imwrite()` |
| 색상 변환 | BGR ↔ RGB, 그레이스케일, HSV 등 | `cv2.cvtColor()` |
| 기하학적 변환 | 크기 조절, 회전, 뒤집기 | `cv2.resize()`, `cv2.flip()` |
| 필터링/블러링 | 노이즈 제거, 이미지 스무딩 | `cv2.GaussianBlur()` |
| 에지 검출 | 경계선 추출 | `cv2.Canny()` |
| 윤곽선 검출 | 객체의 외곽선 추출 | `cv2.findContours()` |
| 도형/텍스트 | 이미지 위에 그리기 | `cv2.line()`, `cv2.putText()` |
| 임계처리 | 이진화(흑백 분리) | `cv2.threshold()` |

### OpenCV 이미지의 본질

OpenCV에서 이미지는 **NumPy 배열(ndarray)**로 표현됨

```
컬러 이미지 (높이 x 너비 x 채널)
  shape = (480, 640, 3)   ← 480행 x 640열 x 3채널(B, G, R)
  dtype = uint8           ← 각 픽셀값: 0~255 (8비트 부호 없는 정수)

그레이스케일 이미지 (높이 x 너비)
  shape = (480, 640)      ← 480행 x 640열, 채널 없음
  dtype = uint8           ← 0(검정) ~ 255(흰색)
```

> **핵심**: OpenCV의 이미지 = NumPy 배열 → NumPy의 인덱싱, 슬라이싱, 연산을 그대로 활용 가능

## 1. 컴퓨터 비전 처리 단계

컴퓨터 비전은 **카메라로 촬영한 이미지를 컴퓨터가 이해하고 분석**하는 기술

### 처리 파이프라인

```
이미지 획득 → 전처리 → 특징 추출 → 고수준 처리 → 결과 출력
(카메라/파일)  (노이즈 제거)  (에지, 윤곽선)  (분류, 탐지)  (판단/시각화)
```

| 단계 | 설명 | OpenCV 활용 |
|------|------|------------|
| **이미지 획득** | 카메라, 파일, URL 등에서 이미지 로드 | `cv2.imread()`, `cv2.VideoCapture()` |
| **전처리** | 노이즈 제거, 크기 조절, 색상 변환 등 | `cv2.GaussianBlur()`, `cv2.resize()` |
| **특징 추출** | 에지, 윤곽선, 코너 등 핵심 특징 검출 | `cv2.Canny()`, `cv2.findContours()` |
| **고수준 처리** | 객체 인식, 분류, 추적 등 의미 파악 | 딥러닝 모델 연동 |

## 3. 라이브러리 불러오기

OpenCV 작업에 필요한 핵심 라이브러리를 불러옴
VSCode 노트북에서는 **matplotlib**로 이미지를 표시함 (Colab의 `cv2_imshow` 대신)

````python
# ──────────────────────────────────────────────
# 필수 라이브러리 불러오기
# ──────────────────────────────────────────────
import cv2                        # OpenCV: 컴퓨터 비전 (이미지 읽기/변환/필터링 등)
import numpy as np                # NumPy: 수치 배열 연산 (이미지 = ndarray)
import matplotlib.pyplot as plt   # matplotlib: 이미지를 노트북에 표시 (VSCode용)
from PIL import Image             # Pillow: Python 이미징 라이브러리 (간단한 이미지 처리)
import requests                   # requests: HTTP 요청 (URL에서 이미지 다운로드)
import os                         # os: 파일/디렉토리 경로 처리

# 버전 확인 출력
print(f"OpenCV 버전: {cv2.__version__}")
print(f"NumPy 버전: {np.__version__}")
````

````python
# ──────────────────────────────────────────────
# 이미지 표시 헬퍼 함수 정의
#   VSCode 노트북에서는 cv2.imshow() 대신 matplotlib 사용
#   이 함수를 이후 모든 셀에서 재사용
# ──────────────────────────────────────────────
def imshow(title, image):
    """OpenCV 이미지를 matplotlib로 표시하는 헬퍼 함수"""
    plt.figure(figsize=(6, 4))           # 6x4 인치 크기의 그림 생성
    if len(image.shape) == 3:            # 컬러 이미지(3채널)인 경우
        # OpenCV는 BGR 순서 → matplotlib는 RGB 순서이므로 변환 필요
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    else:                                # 그레이스케일 이미지(2차원)인 경우
        plt.imshow(image, cmap='gray')   # 회색조 컬러맵으로 표시
    plt.title(title)                     # 이미지 제목 설정
    plt.axis('off')                      # 축 눈금 숨기기
    plt.tight_layout()                   # 레이아웃 자동 조정
    plt.show()                           # 화면에 출력

def imshow_compare(titles, images, figsize=(12, 4)):
    """여러 이미지를 나란히 비교하는 함수"""
    n = len(images)                      # 표시할 이미지 개수
    fig, axes = plt.subplots(1, n, figsize=figsize)  # 1행 n열 서브플롯 생성
    if n == 1:                           # 이미지가 1개면 axes를 리스트로 변환
        axes = [axes]
    for ax, title, img in zip(axes, titles, images):  # 각 서브플롯에 이미지 표시
        if len(img.shape) == 3:          # 컬러 이미지 → BGR→RGB 변환
            ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        else:                            # 그레이스케일 이미지
            ax.imshow(img, cmap='gray')
        ax.set_title(title)              # 서브플롯 제목
        ax.axis('off')                   # 축 눈금 숨기기
    plt.tight_layout()                   # 레이아웃 자동 조정
    plt.show()                           # 화면에 출력

print("헬퍼 함수 정의 완료: imshow(), imshow_compare()")
````

## 4. 이미지 읽기

이미지를 메모리로 불러오는 방법 — **파일**, **URL**, **직접 생성** 3가지

### `cv2.imread()` 주요 매개변수

| 매개변수 | 설명 | 값 |
|---------|------|-----|
| `filename` | 이미지 파일 경로 | `'image.jpg'` |
| `flags` | 읽기 모드 | `cv2.IMREAD_COLOR` (기본, 컬러) |
| | | `cv2.IMREAD_GRAYSCALE` (그레이스케일) |
| | | `cv2.IMREAD_UNCHANGED` (알파 채널 포함) |

> **주의**: `cv2.imread()`는 파일이 없어도 **에러 없이 None을 반환**
> → 반드시 읽기 성공 여부를 확인하는 습관이 중요

````python
# ──────────────────────────────────────────────
# 방법 1: 샘플 이미지 직접 생성 (패키지 없이 바로 실행 가능)
#   NumPy로 컬러 이미지를 만들어 파일로 저장
# ──────────────────────────────────────────────
# 300x400 픽셀, 3채널(BGR) 검정색 이미지 생성 (값 0 = 검정)
sample = np.zeros((300, 400, 3), dtype=np.uint8)

# 색상 영역 채우기 (OpenCV는 BGR 순서!)
sample[0:100, :] = [255, 0, 0]       # 상단 1/3 전체: 파란색 (B=255, G=0, R=0)
sample[100:200, :] = [0, 255, 0]     # 중단 1/3 전체: 초록색 (B=0, G=255, R=0)
sample[200:300, :] = [0, 0, 255]     # 하단 1/3 전체: 빨간색 (B=0, G=0, R=255)
sample[100:200, 125:275] = [255, 255, 255]  # 가운데에 흰색 사각형

# 이미지를 JPEG 파일로 저장
cv2.imwrite('sample_image.jpg', sample)
print("sample_image.jpg 저장 완료")

# 저장한 이미지를 다시 읽어서 확인
image = cv2.imread('sample_image.jpg')  # BGR 컬러로 읽기
if image is not None:
    print(f"이미지 읽기 성공! shape: {image.shape}")  # (높이, 너비, 채널)
else:
    print("이미지 읽기 실패 — 파일 경로를 확인하세요")

# 헬퍼 함수로 이미지 표시 (내부에서 BGR→RGB 변환)
imshow('샘플 이미지 (BGR → RGB 변환하여 표시)', image)
````

````python
# ──────────────────────────────────────────────
# 방법 2: URL에서 이미지 다운로드하여 읽기
#   requests로 다운로드 → numpy 배열 변환 → cv2.imdecode()
# ──────────────────────────────────────────────
# 테스트 이미지 URL
image_url = 'https://filmfare.wwmindia.com/content/2025/jun/thubtsallyouneedtoknow1749639776.jpg'

response = requests.get(image_url)         # HTTP GET 요청으로 이미지 다운로드
if response.status_code == 200:            # HTTP 200 = 성공
    # 응답 바이트 데이터를 numpy 배열로 변환
    image_array = np.asarray(bytearray(response.content), dtype=np.uint8)
    # numpy 배열을 OpenCV 이미지로 디코딩 (컬러 모드)
    image_url_loaded = cv2.imdecode(image_array, cv2.IMREAD_COLOR)
    print(f"URL 이미지 shape: {image_url_loaded.shape}")
    imshow('URL에서 다운로드한 이미지', image_url_loaded)  # 이미지 표시
else:
    print(f"다운로드 실패 (HTTP {response.status_code})")
````

````python
# ──────────────────────────────────────────────
# 방법 3: 로컬 파일에서 읽기 + 그레이스케일 읽기
# ──────────────────────────────────────────────
image_color = cv2.imread('sample_image.jpg')
image_gray = cv2.imread('sample_image.jpg', cv2.IMREAD_GRAYSCALE)

print(f"컬러 이미지 shape: {image_color.shape}")       # (높이, 너비, 3)
print(f"그레이스케일 shape: {image_gray.shape}")        # (높이, 너비)

imshow_compare(['컬러 (BGR→RGB)', '그레이스케일'], [image_color, image_gray], figsize=(10, 4))
````

## 6. 색상 변환

OpenCV는 이미지를 **BGR** 순서로 읽음 (일반적인 RGB와 반대!)
다양한 색상 공간(Color Space) 간 변환이 가능

### 주요 색상 공간

| 색상 공간 | 설명 | 변환 코드 |
|-----------|------|----------|
| **BGR** | OpenCV 기본 (Blue, Green, Red) | — |
| **RGB** | 일반적인 색상 순서 (matplotlib, PIL) | `cv2.COLOR_BGR2RGB` |
| **GRAY** | 그레이스케일 (밝기만) | `cv2.COLOR_BGR2GRAY` |
| **HSV** | 색상(Hue), 채도(Saturation), 명도(Value) | `cv2.COLOR_BGR2HSV` |

### BGR vs RGB 순서 문제

```
OpenCV (BGR):     B=255, G=0,   R=0   → 파란색
matplotlib (RGB): R=0,   G=0,   B=255 → 파란색

같은 이미지라도 변환 없이 표시하면 빨강↔파랑이 바뀌어 보임!
```

````python
# ──────────────────────────────────────────────
# 색상 변환: cv2.cvtColor()
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

print(f"원본(BGR) shape: {image.shape}")
print(f"그레이스케일 shape: {gray.shape}")
print(f"HSV shape: {hsv.shape}")

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('원본 (BGR→RGB 변환)')
axes[1].imshow(image)    # BGR 그대로 → 색상 반전!
axes[1].set_title('BGR 그대로 표시 (색상 반전!)')
axes[2].imshow(gray, cmap='gray')
axes[2].set_title('그레이스케일')
axes[3].imshow(hsv)
axes[3].set_title('HSV')
for ax in axes:
    ax.axis('off')
plt.tight_layout()
plt.show()
````

## 7. 크기 조절과 기하학적 변환

이미지의 **크기, 각도, 방향**을 변경하는 기본 변환

| 함수 | 설명 | 핵심 매개변수 |
|------|------|-------------|
| `cv2.resize()` | 크기 조절 | `(width, height)` |
| `cv2.getRotationMatrix2D()` + `cv2.warpAffine()` | 임의 각도 회전 | `center, angle, scale` |
| `cv2.rotate()` | 90도 단위 회전 | `cv2.ROTATE_90_CLOCKWISE` 등 |
| `cv2.flip()` | 뒤집기 | `flipCode`: 1=좌우, 0=상하, -1=둘 다 |

````python
# ──────────────────────────────────────────────
# 크기 조절 (Resize)
#   cv2.resize(이미지, (너비, 높이))
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')
resized_fixed = cv2.resize(image, (200, 150))          # 고정 크기
resized_half = cv2.resize(image, None, fx=0.5, fy=0.5) # 50% 축소

print(f"원본: {image.shape}")
print(f"고정 크기: {resized_fixed.shape}")
print(f"50% 축소: {resized_half.shape}")

imshow_compare(['원본', '200x150 고정', '50% 축소'], [image, resized_fixed, resized_half], figsize=(14, 4))
````

````python
# ──────────────────────────────────────────────
# 회전 (Rotation)
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')
rows, cols = image.shape[:2]

# 방법 1: 임의 각도 회전
M = cv2.getRotationMatrix2D(center=(cols/2, rows/2), angle=45, scale=1.0)
rotated_45 = cv2.warpAffine(image, M, (cols, rows))

# 방법 2: 90도 단위 회전
rotated_90 = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
rotated_180 = cv2.rotate(image, cv2.ROTATE_180)

imshow_compare(['원본', '45도 회전', '90도 (시계방향)', '180도'],
               [image, rotated_45, rotated_90, rotated_180], figsize=(16, 4))
````

````python
# ──────────────────────────────────────────────
# 뒤집기 (Flip)
#   flipCode: 1=좌우(수평), 0=상하(수직), -1=상하+좌우
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')
h_flipped = cv2.flip(image, flipCode=1)
v_flipped = cv2.flip(image, flipCode=0)
hv_flipped = cv2.flip(image, flipCode=-1)

imshow_compare(['원본', '좌우 뒤집기', '상하 뒤집기', '상하+좌우'],
               [image, h_flipped, v_flipped, hv_flipped], figsize=(16, 4))
````

## 8. 이미지 연산

OpenCV 이미지는 NumPy 배열이므로 **픽셀 단위 연산**이 가능

| 연산 | 함수 | 설명 |
|------|------|------|
| 픽셀 값 변경 | `image[y, x] = [B, G, R]` | 특정 좌표의 색상 변경 |
| 관심 영역(ROI) | `image[y1:y2, x1:x2]` | 이미지의 일부 영역 추출 |
| 더하기 | `cv2.add(img1, img2)` | 픽셀 값 합산 (밝아짐) |
| 차이 | `cv2.absdiff(img1, img2)` | 픽셀 값 차이 (변화 감지) |
| 가중 합성 | `cv2.addWeighted(img1, a, img2, b, g)` | 투명도 조절 합성 |

````python
# ──────────────────────────────────────────────
# 픽셀 값 접근과 변경
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg').copy()
print(f"(0, 0) 픽셀 값(BGR): {image[0, 0]}")

image[50, 50] = [255, 255, 255]
image[100:150, 100:150] = [255, 255, 255]  # 50x50 흰색 사각형

imshow('픽셀 값 변경 결과', image)
````

````python
# ──────────────────────────────────────────────
# ROI(Region of Interest): 관심 영역 추출
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')
roi = image[50:200, 100:300]

print(f"원본: {image.shape},  ROI: {roi.shape}")
imshow_compare(['원본 이미지', 'ROI (잘라낸 영역)'], [image, roi], figsize=(10, 4))
````

````python
# ──────────────────────────────────────────────
# 이미지 더하기와 차이
# ──────────────────────────────────────────────
img1 = np.zeros((200, 300, 3), dtype=np.uint8)
img1[50:150, 50:150] = [255, 0, 0]      # 파란색 사각형

img2 = np.zeros((200, 300, 3), dtype=np.uint8)
img2[75:175, 100:200] = [0, 0, 255]     # 빨간색 사각형

added = cv2.add(img1, img2)              # 더하기: 겹치는 부분 밝아짐
diff = cv2.absdiff(img1, img2)           # 차이: 다른 부분만 남음
blended = cv2.addWeighted(img1, 0.7, img2, 0.3, 0)  # 가중 합성

imshow_compare(['이미지1', '이미지2', '더하기', '차이', '가중합성(0.7:0.3)'],
               [img1, img2, added, diff, blended], figsize=(20, 4))
````

## 9. 필터링과 블러링 (Filtering & Blurring)

이미지의 **노이즈를 제거**하거나 **부드럽게(스무딩)** 만드는 기법
커널(kernel, 작은 행렬)을 이미지 위에 슬라이딩하면서 주변 픽셀값의 평균/가중평균을 계산

| 함수 | 설명 | 특징 |
|------|------|------|
| `cv2.blur()` | 평균 블러 | 단순 평균, 가장 빠름 |
| `cv2.GaussianBlur()` | 가우시안 블러 | 정규분포 가중치, 자연스러운 흐림 |
| `cv2.medianBlur()` | 중간값 블러 | 소금-후추 노이즈에 효과적 |
| `cv2.bilateralFilter()` | 양방향 필터 | 경계를 보존하면서 노이즈 제거 |

> 커널 크기가 클수록 더 강하게 블러 처리됨 (항상 **홀수**로 지정)

````python
# ──────────────────────────────────────────────
# 블러링(Blurring) 비교
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')

blur_avg = cv2.blur(image, (7, 7))                 # 평균 블러
blur_gauss = cv2.GaussianBlur(image, (7, 7), 0)    # 가우시안 블러
blur_median = cv2.medianBlur(image, 7)              # 중간값 블러
blur_bilateral = cv2.bilateralFilter(image, 9, 75, 75)  # 양방향 필터

imshow_compare(['원본', '평균 블러', '가우시안 블러', '중간값 블러'],
               [image, blur_avg, blur_gauss, blur_median], figsize=(16, 4))
````

## 10. 임계 처리 — 이진화 (Thresholding)

이미지를 **흑(0)과 백(255) 두 값으로 분리**하는 기법
임계값(threshold)보다 크면 흰색, 작으면 검정으로 변환

| 방법 | 함수 | 설명 |
|------|------|------|
| 전역 임계 | `cv2.threshold()` | 고정된 하나의 임계값 사용 |
| 적응형 임계 | `cv2.adaptiveThreshold()` | 영역별로 다른 임계값 자동 계산 |
| Otsu 방법 | `cv2.threshold(..., cv2.THRESH_OTSU)` | 최적 임계값 자동 결정 |

```
원본 픽셀값:  30  128  200  50  180  90
임계값 = 128
결과:          0    0   255   0  255   0
              검정  검정  흰색  검정  흰색  검정
```

````python
# ──────────────────────────────────────────────
# 임계 처리 (Thresholding)
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

thresh_val = 128
_, binary = cv2.threshold(gray, thresh_val, 255, cv2.THRESH_BINARY)
_, binary_inv = cv2.threshold(gray, thresh_val, 255, cv2.THRESH_BINARY_INV)
otsu_val, binary_otsu = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print(f"Otsu가 자동 결정한 임계값: {otsu_val}")

imshow_compare(['그레이스케일', f'이진화 (임계={thresh_val})', '반전 이진화', f'Otsu (임계={otsu_val:.0f})'],
               [gray, binary, binary_inv, binary_otsu], figsize=(16, 4))
````

## 11. 에지 검출 (Edge Detection)

이미지에서 **밝기가 급격히 변하는 지점(경계)**을 찾아내는 기법
객체의 윤곽, 형태 파악의 기초

### Canny 에지 검출 — 가장 널리 사용

```
cv2.Canny(image, threshold1, threshold2)
```

| 매개변수 | 설명 |
|---------|------|
| `image` | 입력 이미지 (그레이스케일 권장) |
| `threshold1` | 최소 임계값 — 이보다 작은 에지 강도는 무시 |
| `threshold2` | 최대 임계값 — 이보다 큰 에지 강도는 확실한 에지 |

> 두 임계값 사이의 에지는 **주변 상황을 보고** 에지 포함 여부를 결정 (히스테리시스)

````python
# ──────────────────────────────────────────────
# Canny 에지 검출
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

edges_low = cv2.Canny(gray, 30, 100)     # 낮은 임계값 → 에지 많이 검출
edges_mid = cv2.Canny(gray, 50, 150)     # 중간 임계값
edges_high = cv2.Canny(gray, 100, 200)   # 높은 임계값 → 강한 에지만

imshow_compare(['원본', 'Canny (30, 100)', 'Canny (50, 150)', 'Canny (100, 200)'],
               [gray, edges_low, edges_mid, edges_high], figsize=(16, 4))
````

## 12. 윤곽선 검출 (Contour Detection)

**이진화된 이미지에서 객체의 외곽선**을 찾아내는 기법
객체의 형태 분석, 개수 세기, 면적 계산 등에 활용

### 처리 순서

```
컬러 이미지 → 그레이스케일 → 이진화(threshold) → 윤곽선 검출(findContours)
```

| 함수 | 설명 |
|------|------|
| `cv2.findContours()` | 윤곽선 좌표 추출 |
| `cv2.drawContours()` | 이미지 위에 윤곽선 그리기 |

| 매개변수 | 설명 |
|---------|------|
| `cv2.RETR_EXTERNAL` | 가장 바깥쪽 윤곽선만 검출 |
| `cv2.RETR_TREE` | 모든 윤곽선 (계층 구조 포함) |
| `cv2.CHAIN_APPROX_SIMPLE` | 윤곽선 좌표를 압축 저장 |

````python
# ──────────────────────────────────────────────
# 윤곽선 검출 (Contour Detection)
# ──────────────────────────────────────────────
canvas = np.zeros((300, 400, 3), dtype=np.uint8)
cv2.rectangle(canvas, (50, 50), (150, 150), (255, 255, 255), -1)
cv2.circle(canvas, (300, 150), 70, (255, 255, 255), -1)

gray = cv2.cvtColor(canvas, cv2.COLOR_BGR2GRAY)
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

contours, hierarchy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
print(f"검출된 윤곽선 수: {len(contours)}")

result = canvas.copy()
cv2.drawContours(result, contours, -1, (0, 255, 0), 2)

imshow_compare(['원본 도형', '이진화', '윤곽선 검출 결과'],
               [canvas, binary, result], figsize=(14, 4))
````

## 13. 히스토그램 평활화 (Histogram Equalization)

이미지의 **픽셀 밝기 분포를 균등하게 재분배**하여 **대비(contrast)를 향상**시키는 기법
너무 어둡거나 너무 밝아 세부 정보가 묻힌 이미지를 복원할 때 효과적

```
평활화 전:                    평활화 후:
빈도 |                       빈도 |
     |  ##                        | ## ## ## ## ##
     |  ## #                      | ## ## ## ## ##
     └──────→ 밝기              └──────────────→ 밝기
     (한곳에 몰려 있음)         (고르게 분포)
```

````python
# ──────────────────────────────────────────────
# 히스토그램 평활화
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
equalized = cv2.equalizeHist(gray)

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes[0, 0].imshow(gray, cmap='gray')
axes[0, 0].set_title('원본 그레이스케일')
axes[0, 0].axis('off')
axes[0, 1].imshow(equalized, cmap='gray')
axes[0, 1].set_title('평활화 적용 후')
axes[0, 1].axis('off')
axes[1, 0].hist(gray.ravel(), 256, [0, 256], color='steelblue', alpha=0.7)
axes[1, 0].set_title('원본 히스토그램')
axes[1, 0].set_xlabel('픽셀 값')
axes[1, 1].hist(equalized.ravel(), 256, [0, 256], color='coral', alpha=0.7)
axes[1, 1].set_title('평활화 후 히스토그램')
axes[1, 1].set_xlabel('픽셀 값')
plt.tight_layout()
plt.show()
````

## 14. 이미지에 도형과 텍스트 그리기

이미지 위에 **선, 사각형, 원, 텍스트** 등을 그려 시각적 표시 가능
객체 검출 결과를 표시하거나, 이미지에 주석을 달 때 사용

| 함수 | 설명 | 핵심 매개변수 |
|------|------|-------------|
| `cv2.line()` | 직선 | 시작점, 끝점, 색상, 두께 |
| `cv2.rectangle()` | 사각형 | 좌상단, 우하단, 색상, 두께 |
| `cv2.circle()` | 원 | 중심, 반지름, 색상, 두께(-1이면 채움) |
| `cv2.putText()` | 텍스트 | 텍스트, 위치, 폰트, 크기, 색상 |

> **주의**: 그리기 함수는 **원본 이미지를 직접 수정**함
> 원본 보존이 필요하면 `.copy()`로 복사 후 사용

````python
# ──────────────────────────────────────────────
# 도형 그리기
# ──────────────────────────────────────────────
canvas = np.zeros((400, 500, 3), dtype=np.uint8)

cv2.line(canvas, (50, 50), (450, 50), (255, 255, 255), 2)          # 직선
cv2.rectangle(canvas, (50, 100), (200, 250), (0, 255, 0), 3)       # 사각형 (테두리)
cv2.rectangle(canvas, (250, 100), (400, 250), (0, 255, 0), -1)     # 사각형 (채움)
cv2.circle(canvas, (125, 330), 50, (0, 0, 255), 2)                 # 원 (테두리)
cv2.circle(canvas, (325, 330), 50, (255, 0, 0), -1)                # 원 (채움)
cv2.putText(canvas, 'OpenCV Drawing!', (100, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

imshow('도형과 텍스트 그리기', canvas)
````

````python
# ──────────────────────────────────────────────
# 기존 이미지 위에 도형 그리기 (객체 표시 시뮬레이션)
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg').copy()  # 원본 보존을 위해 복사!

cv2.rectangle(image, (50, 50), (200, 200), (0, 255, 0), 2)     # 바운딩 박스
cv2.circle(image, (300, 150), 30, (0, 0, 255), -1)             # 포인트 마커
cv2.line(image, (0, 0), (400, 300), (255, 0, 0), 2)            # 직선
cv2.putText(image, 'Region', (55, 45),
            cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

imshow('이미지 위에 도형 추가', image)
````

## 15. 이미지 저장

처리된 이미지를 **파일로 저장**하는 방법

| 확장자 | 형식 | 특징 |
|--------|------|------|
| `.jpg` / `.jpeg` | JPEG | 손실 압축, 파일 크기 작음, 사진에 적합 |
| `.png` | PNG | 무손실 압축, 투명도(알파 채널) 지원 |
| `.bmp` | BMP | 무압축, 파일 크기 큼, 처리 빠름 |

> `cv2.imwrite()`는 성공 시 `True`, 실패 시 `False` 반환

````python
# ──────────────────────────────────────────────
# 이미지 저장
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 50, 150)

success1 = cv2.imwrite('output_gray.jpg', gray)
print(f"그레이스케일 저장: {'성공' if success1 else '실패'}")

success2 = cv2.imwrite('output_edges.png', edges)
print(f"에지 검출 결과 저장: {'성공' if success2 else '실패'}")

saved_files = [f for f in os.listdir('.') if f.startswith('output_')]
print(f"저장된 파일: {saved_files}")
````

## 16. 이미지 딥러닝 전처리 기법 요약

딥러닝 모델에 이미지를 입력하기 전에 수행하는 **표준 전처리 단계**
모델의 성능과 학습 효율에 직접적인 영향을 미침

| 단계 | 기법 | OpenCV 함수 | 목적 |
|------|------|------------|------|
| 1 | 크기 조절 | `cv2.resize(img, (224, 224))` | 모델 입력 크기 통일 |
| 2 | 색상 변환 | `cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)` | 채널 수 맞추기 |
| 3 | 정규화 | `cv2.normalize()` 또는 `img / 255.0` | 픽셀값 0~1 범위로 조정 |
| 4 | 데이터 증강 | `cv2.flip()`, `cv2.rotate()` | 학습 데이터 다양성 확보 |
| 5 | 노이즈 추가 | `np.random` 활용 | 모델 강건성 향상 |
| 6 | 히스토그램 평활화 | `cv2.equalizeHist()` | 대비 개선 |

````python
# ──────────────────────────────────────────────
# 딥러닝 전처리 파이프라인 예시
# ──────────────────────────────────────────────
image = cv2.imread('sample_image.jpg')      # 원본 이미지 읽기

# 1) 크기 조절: 대부분 딥러닝 모델은 고정 입력 크기 요구 (예: 224x224)
resized = cv2.resize(image, (224, 224))     # (너비, 높이) 순서로 지정
print(f"크기 조절: {image.shape} -> {resized.shape}")

# 2) 색상 변환: BGR → 그레이스케일 (채널 수 줄이기)
gray = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)

# 3) 정규화: 픽셀값을 0~1 범위로 변환 (딥러닝 학습 효율 향상)
normalized = resized.astype(np.float32) / 255.0  # uint8(0~255) → float32(0~1)
print(f"정규화 전 범위: {resized.min()} ~ {resized.max()}")
print(f"정규화 후 범위: {normalized.min():.1f} ~ {normalized.max():.1f}")

# 4) 데이터 증강: 학습 데이터 다양성 확보
augmented_flip = cv2.flip(resized, 1)       # 좌우 뒤집기 (flipCode=1)
augmented_rot = cv2.rotate(resized, cv2.ROTATE_90_CLOCKWISE)  # 시계방향 90도 회전

# 5) 노이즈 추가: 모델 강건성(robustness) 향상
noisy = resized.copy()                       # 원본 복사
# 가우시안 노이즈 생성 (평균=0, 표준편차=25)
noise = np.random.normal(0, 25, resized.shape)
# 노이즈를 더한 후 0~255 범위로 클리핑하여 uint8 변환
noisy = np.clip(noisy.astype(np.float32) + noise, 0, 255).astype(np.uint8)

# 4가지 증강 결과 비교 표시
imshow_compare(['원본 (224x224)', '좌우 뒤집기', '90도 회전', '노이즈 추가'],
               [resized, augmented_flip, augmented_rot, noisy], figsize=(16, 4))
````

## OpenCV 핵심 함수 요약 (Quick Reference)

| 카테고리 | 함수 | 설명 |
|---------|------|------|
| **읽기/쓰기** | `cv2.imread(path)` | 이미지 파일 읽기 |
| | `cv2.imdecode(buf, flag)` | 바이트 배열에서 이미지 디코딩 |
| | `cv2.imwrite(path, img)` | 이미지 파일 저장 |
| **색상 변환** | `cv2.cvtColor(img, code)` | 색상 공간 변환 (BGR↔RGB, GRAY, HSV) |
| **크기/변환** | `cv2.resize(img, (w, h))` | 크기 조절 |
| | `cv2.rotate(img, code)` | 90도 단위 회전 |
| | `cv2.flip(img, code)` | 뒤집기 (0=상하, 1=좌우, -1=둘다) |
| | `cv2.warpAffine(img, M, size)` | 어파인 변환 (임의 회전 등) |
| **필터링** | `cv2.GaussianBlur(img, ksize, sigma)` | 가우시안 블러 |
| | `cv2.medianBlur(img, ksize)` | 중간값 블러 |
| | `cv2.bilateralFilter(img, d, sc, ss)` | 양방향 필터 |
| **임계처리** | `cv2.threshold(img, thresh, max, type)` | 전역 임계 처리 |
| | `cv2.adaptiveThreshold()` | 적응형 임계 처리 |
| **에지/윤곽선** | `cv2.Canny(img, t1, t2)` | Canny 에지 검출 |
| | `cv2.findContours(img, mode, method)` | 윤곽선 검출 |
| | `cv2.drawContours(img, contours, idx, color, thick)` | 윤곽선 그리기 |
| **히스토그램** | `cv2.equalizeHist(img)` | 히스토그램 평활화 |
| **그리기** | `cv2.line(img, pt1, pt2, color, thick)` | 직선 |
| | `cv2.rectangle(img, pt1, pt2, color, thick)` | 사각형 |
| | `cv2.circle(img, center, r, color, thick)` | 원 |
| | `cv2.putText(img, text, pos, font, scale, color)` | 텍스트 |
| **연산** | `cv2.add(img1, img2)` | 이미지 더하기 |
| | `cv2.absdiff(img1, img2)` | 이미지 차이 |
| | `cv2.addWeighted(img1, a, img2, b, g)` | 가중 합성 |
| **정규화** | `cv2.normalize(img, None, a, b, type)` | 값 범위 정규화 |
| **표시 (VSCode)** | `plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))` | matplotlib로 표시 |
